```python
# %% [markdown]
# # üîÆ Sliding Window Prediction
#
# ÌïôÏäµÎêú Î™®Îç∏Î°ú ÎåÄÌòï Ïù¥ÎØ∏ÏßÄ (4096√ó4096)Î•º Ïä¨ÎùºÏù¥Îî© ÏúàÎèÑÏö∞ Î∞©ÏãùÏúºÎ°ú ÏòàÏ∏°Ìï©ÎãàÎã§.
#
# ## üìã Î™©Ï∞®
# 1. ÌôòÍ≤Ω ÏÑ§Ï†ï Î∞è ÏûÑÌè¨Ìä∏
# 2. ÌååÎùºÎØ∏ÌÑ∞ ÏÑ§Ï†ï
# 3. Î™®Îç∏ Î°úÎìú
# 4. Ïä¨ÎùºÏù¥Îî© ÏúàÎèÑÏö∞ ÏòàÏ∏°
# 5. Í≤∞Í≥º ÏãúÍ∞ÅÌôî Î∞è Ï†ÄÏû•

# %% [markdown]
# ## 1. ÌôòÍ≤Ω ÏÑ§Ï†ï Î∞è ÏûÑÌè¨Ìä∏

# %%
import torch
import numpy as np
import cv2
import matplotlib.pyplot as plt
from pathlib import Path
import sys
from tqdm import tqdm

# PyTorch ÏΩîÎìú Í≤ΩÎ°ú Ï∂îÍ∞Ä
sys.path.append('pytorch_codes')

from models import ForwardPhaseUNet, MultiScalePhaseUNet, PhaseAmplitudeUNet

# GPU ÌôïÏù∏
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(f"‚úÖ ÌôòÍ≤Ω ÏÑ§Ï†ï ÏôÑÎ£å!")
print(f"   Device: {device}")
print(f"   PyTorch Î≤ÑÏ†Ñ: {torch.__version__}")

# %% [markdown]
# ## 2. ÌååÎùºÎØ∏ÌÑ∞ ÏÑ§Ï†ï

# %%
# ==================== ÏûÖÎ†• ÌååÎùºÎØ∏ÌÑ∞ ====================
INPUT_MASK_PATH = 'data/forward_intensity/inputs/sample_0000.png'  # ÏòàÏ∏°Ìï† Ïù¥ÎØ∏ÏßÄ
OUTPUT_DIR = 'predictions'                                      # Ï∂úÎ†• ÎîîÎ†âÌÜ†Î¶¨

# ==================== Î™®Îç∏ ÌååÎùºÎØ∏ÌÑ∞ (ÌïôÏäµ ÏãúÏôÄ ÎèôÏùºÌï¥Ïïº Ìï®) ====================
MODEL_TYPE = 'basic'            # 'basic', 'multiscale', 'phase_amplitude'
LAYER_NUM = 5                   # U-Net Î†àÏù¥Ïñ¥ Ïàò
BASE_FEATURES = 64              # Í∏∞Î≥∏ feature Ïàò
DROPOUT_RATE = 0.2              # Dropout ÎπÑÏú®
USE_BATCHNORM = True            # BatchNorm ÏÇ¨Ïö© Ïó¨Î∂Ä
EXPERIMENT_NAME = f'forward_phase_{MODEL_TYPE}_tiles'
CHECKPOINT_PATH = f'checkpoints/{EXPERIMENT_NAME}/best_model.pth'  # ÌïôÏäµÎêú Î™®Îç∏

# ==================== Ïä¨ÎùºÏù¥Îî© ÏúàÎèÑÏö∞ ÌååÎùºÎØ∏ÌÑ∞ ====================
TILE_SIZE = 256                 # ÌÉÄÏùº ÌÅ¨Í∏∞ (ÌïôÏäµ ÏãúÏôÄ ÎèôÏùºÌï¥Ïïº Ìï®)
STRIDE = 64                     # Ïä¨ÎùºÏù¥Îî© stride (ÏûëÏùÑÏàòÎ°ù Îçî Ï†ïÌôï, ÎäêÎ¶º)

# Ï∂úÎ†• ÎîîÎ†âÌÜ†Î¶¨ ÏÉùÏÑ±
Path(OUTPUT_DIR).mkdir(parents=True, exist_ok=True)

print("‚úÖ ÌååÎùºÎØ∏ÌÑ∞ ÏÑ§Ï†ï ÏôÑÎ£å!")
print(f"\nüìä ÏòàÏ∏° ÏÑ§Ï†ï:")
print(f"   ÏûÖÎ†• Ïù¥ÎØ∏ÏßÄ: {INPUT_MASK_PATH}")
print(f"   Î™®Îç∏ ÌÉÄÏûÖ: {MODEL_TYPE}")
print(f"   Ïã§Ìóò Ïù¥Î¶Ñ: {EXPERIMENT_NAME}")
print(f"   Ï≤¥ÌÅ¨Ìè¨Ïù∏Ìä∏: {CHECKPOINT_PATH}")
print(f"   ÌÉÄÏùº ÌÅ¨Í∏∞: {TILE_SIZE}√ó{TILE_SIZE}")
print(f"   Stride: {STRIDE}")
print(f"   Device: {device}")

# %% [markdown]
# ## 3. ÏûÖÎ†• Ïù¥ÎØ∏ÏßÄ Î°úÎìú

# %%
print("\nüìÇ ÏûÖÎ†• Ïù¥ÎØ∏ÏßÄ Î°úÎî© Ï§ë...")

input_mask = cv2.imread(INPUT_MASK_PATH, cv2.IMREAD_GRAYSCALE)

if input_mask is None:
    raise ValueError(f"Failed to load image: {INPUT_MASK_PATH}")

h, w = input_mask.shape

print(f"‚úÖ ÏûÖÎ†• Ïù¥ÎØ∏ÏßÄ Î°úÎìú ÏôÑÎ£å!")
print(f"   ÌÅ¨Í∏∞: {w}√ó{h}")
print(f"   Fill ratio: {np.sum(input_mask > 128) / input_mask.size * 100:.1f}%")

# ÏûÖÎ†• Ïù¥ÎØ∏ÏßÄ ÏãúÍ∞ÅÌôî
plt.figure(figsize=(10, 10))
plt.imshow(input_mask, cmap='gray')
plt.title(f'Input Mask\n{w}√ó{h}')
plt.axis('off')
plt.tight_layout()
plt.show()

# %% [markdown]
# ## 4. Î™®Îç∏ Î°úÎìú

# %%
print("\nüì• Î™®Îç∏ Î°úÎî© Ï§ë...")

# Î™®Îç∏ ÏÉùÏÑ± (ÌïôÏäµ ÏãúÏôÄ ÎèôÏùºÌïú ÌååÎùºÎØ∏ÌÑ∞ ÏÇ¨Ïö©)
if MODEL_TYPE == 'basic':
    model = ForwardPhaseUNet(
        in_channels=1,
        out_channels=1,
        layer_num=LAYER_NUM,
        base_features=BASE_FEATURES,
        dropout_rate=DROPOUT_RATE,
        output_activation='linear',
        use_batchnorm=USE_BATCHNORM
    )
elif MODEL_TYPE == 'multiscale':
    model = MultiScalePhaseUNet(
        in_channels=1,
        out_channels=1,
        layer_num=LAYER_NUM,
        base_features=BASE_FEATURES,
        dropout_rate=DROPOUT_RATE,
        use_batchnorm=USE_BATCHNORM
    )
elif MODEL_TYPE == 'phase_amplitude':
    model = PhaseAmplitudeUNet(
        in_channels=1,
        layer_num=LAYER_NUM,
        base_features=BASE_FEATURES,
        dropout_rate=DROPOUT_RATE,
        use_batchnorm=USE_BATCHNORM
    )

# Ï≤¥ÌÅ¨Ìè¨Ïù∏Ìä∏ Î°úÎìú
checkpoint = torch.load(CHECKPOINT_PATH, map_location=device)
model.load_state_dict(checkpoint['model_state_dict'])
model = model.to(device)
model.eval()

print(f"‚úÖ Î™®Îç∏ Î°úÎìú ÏôÑÎ£å!")
print(f"   Î™®Îç∏ ÌÉÄÏûÖ: {MODEL_TYPE}")
print(f"   Epoch: {checkpoint.get('epoch', 'unknown')}")

# Val loss Ï∂úÎ†• (ÏûàÎäî Í≤ΩÏö∞ÏóêÎßå)
if 'val_loss' in checkpoint:
    print(f"   Val Loss: {checkpoint['val_loss']:.6f}")

# %% [markdown]
# ## 5. Ïä¨ÎùºÏù¥Îî© ÏúàÎèÑÏö∞ ÏòàÏ∏°

# %%
print("\n" + "="*80)
print("üîç Ïä¨ÎùºÏù¥Îî© ÏúàÎèÑÏö∞ ÏòàÏ∏° ÏãúÏûë")
print("="*80)

# ÏòàÏ∏° Îßµ Î∞è Ïπ¥Ïö¥Ìä∏ Îßµ Ï¥àÍ∏∞Ìôî
prediction_map = np.zeros((h, w), dtype=np.float32)
count_map = np.zeros((h, w), dtype=np.int32)

# ÌÉÄÏùº Í∞úÏàò Í≥ÑÏÇ∞
n_tiles_y = (h - TILE_SIZE) // STRIDE + 1
n_tiles_x = (w - TILE_SIZE) // STRIDE + 1
total_tiles = n_tiles_y * n_tiles_x

print(f"\nüìê Ïä¨ÎùºÏù¥Îî© ÏúàÎèÑÏö∞ Ï†ïÎ≥¥:")
print(f"   Ïù¥ÎØ∏ÏßÄ ÌÅ¨Í∏∞: {w}√ó{h}")
print(f"   ÌÉÄÏùº ÌÅ¨Í∏∞: {TILE_SIZE}√ó{TILE_SIZE}")
print(f"   Stride: {STRIDE}")
print(f"   ÌÉÄÏùº Í∞úÏàò: {n_tiles_y}√ó{n_tiles_x} = {total_tiles:,}")
print(f"\nÏßÑÌñâ Ï§ë...\n")

# Ïä¨ÎùºÏù¥Îî© ÏúàÎèÑÏö∞ ÏòàÏ∏°
with torch.no_grad():
    pbar = tqdm(total=total_tiles, desc="Processing tiles")
    
    for i in range(n_tiles_y):
        for j in range(n_tiles_x):
            # ÌÉÄÏùº ÏúÑÏπò Í≥ÑÏÇ∞
            top = i * STRIDE
            left = j * STRIDE
            bottom = min(top + TILE_SIZE, h)
            right = min(left + TILE_SIZE, w)
            
            # Í≤ΩÍ≥Ñ Ï≤òÎ¶¨
            if bottom - top < TILE_SIZE:
                top = max(0, bottom - TILE_SIZE)
            if right - left < TILE_SIZE:
                left = max(0, right - TILE_SIZE)
            
            # ÌÉÄÏùº Ï∂îÏ∂ú
            tile = input_mask[top:bottom, left:right]
            
            # ÌÖêÏÑúÎ°ú Î≥ÄÌôò
            tile_tensor = torch.from_numpy(tile).unsqueeze(0).unsqueeze(0).float()
            tile_tensor = tile_tensor / 255.0  # Normalize to [0, 1]
            tile_tensor = tile_tensor.to(device)
            
            # ÏòàÏ∏°
            output = model(tile_tensor)
            
            # Í≤∞Í≥º Ï∂îÏ∂ú
            pred = output.squeeze().cpu().numpy()
            
            # ÏòàÏ∏° ÎßµÏóê Ï∂îÍ∞Ä
            prediction_map[top:bottom, left:right] += pred
            count_map[top:bottom, left:right] += 1
            
            pbar.update(1)
    
    pbar.close()

# ÌèâÍ∑†Ìôî
count_map = np.maximum(count_map, 1)  # 0ÏúºÎ°ú ÎÇòÎàÑÍ∏∞ Î∞©ÏßÄ
prediction_map = prediction_map / count_map

print(f"\n‚úÖ ÏòàÏ∏° ÏôÑÎ£å!")
print(f"   ÌîΩÏÖÄÎãπ ÌèâÍ∑† ÏòàÏ∏° ÌöüÏàò: {np.mean(count_map):.1f}")
print(f"   ÏµúÏÜå ÏòàÏ∏° ÌöüÏàò: {np.min(count_map)}")
print(f"   ÏµúÎåÄ ÏòàÏ∏° ÌöüÏàò: {np.max(count_map)}")

# %% [markdown]
# ## 6. Í≤∞Í≥º Ï†ÄÏû•

# %%
print("\nüíæ Í≤∞Í≥º Ï†ÄÏû• Ï§ë...")

output_path = Path(OUTPUT_DIR)

# ÏòàÏ∏° ÏúÑÏÉÅÎßµ Ï†ÄÏû•
phase_path = output_path / 'predicted_intensity_map.npy'
np.save(phase_path, prediction_map.astype(np.float32))
print(f"   ‚úì Phase map: {phase_path}")

# Ïπ¥Ïö¥Ìä∏ Îßµ Ï†ÄÏû•
count_path = output_path / 'count_map.npy'
np.save(count_path, count_map.astype(np.int32))
print(f"   ‚úì Count map: {count_path}")

print("\n‚úÖ Í≤∞Í≥º Ï†ÄÏû• ÏôÑÎ£å!")

# %% [markdown]
# ## 7. Í≤∞Í≥º ÏãúÍ∞ÅÌôî

# %%
print("\nüé® Í≤∞Í≥º ÏãúÍ∞ÅÌôî Ï§ë...\n")

fig, axes = plt.subplots(2, 2, figsize=(16, 16))

# ÏûÖÎ†• ÎßàÏä§ÌÅ¨
axes[0, 0].imshow(input_mask, cmap='gray')
axes[0, 0].set_title('Input: Random Pillar Mask', fontsize=14, fontweight='bold')
axes[0, 0].axis('off')

# ÏòàÏ∏° ÏúÑÏÉÅÎßµ
im1 = axes[0, 1].imshow(prediction_map, cmap='hsv', vmin=-np.pi, vmax=np.pi)
axes[0, 1].set_title('Predicted Phase Map', fontsize=14, fontweight='bold')
axes[0, 1].axis('off')
plt.colorbar(im1, ax=axes[0, 1], label='Phase (rad)', fraction=0.046)

# Ïπ¥Ïö¥Ìä∏ Îßµ (overlap Ï†ïÎ≥¥)
im2 = axes[1, 0].imshow(count_map, cmap='viridis')
axes[1, 0].set_title('Prediction Count Map\n(Overlapping Predictions)', fontsize=14, fontweight='bold')
axes[1, 0].axis('off')
plt.colorbar(im2, ax=axes[1, 0], label='Count', fraction=0.046)

# ÏúÑÏÉÅ ÌûàÏä§ÌÜ†Í∑∏Îû®
axes[1, 1].hist(prediction_map.flatten(), bins=50, alpha=0.7, edgecolor='black')
axes[1, 1].set_xlabel('Phase (rad)', fontsize=12)
axes[1, 1].set_ylabel('Count', fontsize=12)
axes[1, 1].set_title('Phase Distribution', fontsize=14, fontweight='bold')
axes[1, 1].grid(True, alpha=0.3)
axes[1, 1].axvline(x=np.mean(prediction_map), color='red', linestyle='--', 
                  linewidth=2, label=f'Mean: {np.mean(prediction_map):.3f}')
axes[1, 1].legend()

plt.tight_layout()

# ÏãúÍ∞ÅÌôî Ï†ÄÏû•
vis_path = output_path / 'prediction_visualization.png'
plt.savefig(vis_path, dpi=300, bbox_inches='tight')
print(f"   ‚úì Visualization: {vis_path}")

plt.show()

print("\n‚úÖ ÏãúÍ∞ÅÌôî ÏôÑÎ£å!")

# %% [markdown]
# ## 8. ÌÜµÍ≥Ñ Ï†ïÎ≥¥

# %%
print("\n" + "="*80)
print("üìä ÏòàÏ∏° ÌÜµÍ≥Ñ")
print("="*80)
print(f"\nÏúÑÏÉÅÎßµ:")
print(f"   ÌèâÍ∑†: {np.mean(prediction_map):.4f} rad ({np.mean(prediction_map)/np.pi:.2f}œÄ)")
print(f"   ÌëúÏ§ÄÌé∏Ï∞®: {np.std(prediction_map):.4f} rad ({np.std(prediction_map)/np.pi:.2f}œÄ)")
print(f"   ÏµúÏÜåÍ∞í: {np.min(prediction_map):.4f} rad ({np.min(prediction_map)/np.pi:.2f}œÄ)")
print(f"   ÏµúÎåÄÍ∞í: {np.max(prediction_map):.4f} rad ({np.max(prediction_map)/np.pi:.2f}œÄ)")
print(f"   Î≤îÏúÑ: {np.max(prediction_map) - np.min(prediction_map):.4f} rad")

print(f"\nÏπ¥Ïö¥Ìä∏ Îßµ:")
print(f"   ÌèâÍ∑† ÏòàÏ∏° ÌöüÏàò: {np.mean(count_map):.1f}")
print(f"   ÏµúÏÜå ÏòàÏ∏° ÌöüÏàò: {np.min(count_map)}")
print(f"   ÏµúÎåÄ ÏòàÏ∏° ÌöüÏàò: {np.max(count_map)}")

print(f"\nÏ∂úÎ†• ÌååÏùº:")
print(f"   üìÅ {output_path}/")
print(f"      ‚îú‚îÄ‚îÄ predicted_intensity_map.npy")
print(f"      ‚îú‚îÄ‚îÄ count_map.npy")
print(f"      ‚îî‚îÄ‚îÄ prediction_visualization.png")

# %% [markdown]
# ## 9. Ground TruthÏôÄ ÎπÑÍµê (ÏÑ†ÌÉùÏÇ¨Ìï≠)
#
# Ground TruthÍ∞Ä ÏûàÎäî Í≤ΩÏö∞ ÎπÑÍµêÌï©ÎãàÎã§.

# %%
# Ground Truth Í≤ΩÎ°ú
gt_path = INPUT_MASK_PATH.replace('inputs', 'outputs').replace('.png', '.npy')

if Path(gt_path).exists():
    print("\nüìä Ground TruthÏôÄ ÎπÑÍµê Ï§ë...")
    
    # Ground Truth Î°úÎìú
    ground_truth = np.load(gt_path)
    
    # ÌÅ¨Í∏∞Í∞Ä Îã§Î•¥Î©¥ Ï°∞Ï†ï
    if ground_truth.shape != prediction_map.shape:
        print(f"   ‚ö†Ô∏è  ÌÅ¨Í∏∞ Î∂àÏùºÏπò: GT {ground_truth.shape} vs Pred {prediction_map.shape}")
        print(f"   Ground TruthÎ•º ÏòàÏ∏° ÌÅ¨Í∏∞Î°ú Î¶¨ÏÇ¨Ïù¥Ï¶àÌï©ÎãàÎã§...")
        from scipy import ndimage
        zoom_factors = (prediction_map.shape[0] / ground_truth.shape[0],
                       prediction_map.shape[1] / ground_truth.shape[1])
        ground_truth = ndimage.zoom(ground_truth, zoom_factors, order=1)
    
    # ÏóêÎü¨ Í≥ÑÏÇ∞
    mae = np.mean(np.abs(prediction_map - ground_truth))
    mse = np.mean((prediction_map - ground_truth)**2)
    rmse = np.sqrt(mse)
    
    print(f"\n   ÏóêÎü¨ Î©îÌä∏Î¶≠:")
    print(f"      MAE:  {mae:.4f} rad ({mae/np.pi:.3f}œÄ)")
    print(f"      MSE:  {mse:.4f}")
    print(f"      RMSE: {rmse:.4f} rad ({rmse/np.pi:.3f}œÄ)")
    
    # ÎπÑÍµê ÏãúÍ∞ÅÌôî
    fig, axes = plt.subplots(2, 2, figsize=(16, 16))
    
    # Ground Truth
    im1 = axes[0, 0].imshow(ground_truth, cmap='hsv', vmin=-np.pi, vmax=np.pi)
    axes[0, 0].set_title('Ground Truth (MEEP)', fontsize=14, fontweight='bold')
    axes[0, 0].axis('off')
    plt.colorbar(im1, ax=axes[0, 0], label='Phase (rad)', fraction=0.046)
    
    # Prediction
    im2 = axes[0, 1].imshow(prediction_map, cmap='hsv', vmin=-np.pi, vmax=np.pi)
    axes[0, 1].set_title('Prediction (U-Net)', fontsize=14, fontweight='bold')
    axes[0, 1].axis('off')
    plt.colorbar(im2, ax=axes[0, 1], label='Phase (rad)', fraction=0.046)
    
    # Error map
    error = np.abs(prediction_map - ground_truth)
    im3 = axes[1, 0].imshow(error, cmap='hot')
    axes[1, 0].set_title(f'Absolute Error\nMAE = {mae:.4f} rad', fontsize=14, fontweight='bold')
    axes[1, 0].axis('off')
    plt.colorbar(im3, ax=axes[1, 0], label='Error (rad)', fraction=0.046)
    
    # Error histogram
    axes[1, 1].hist(error.flatten(), bins=50, alpha=0.7, edgecolor='black')
    axes[1, 1].set_xlabel('Absolute Error (rad)', fontsize=12)
    axes[1, 1].set_ylabel('Count', fontsize=12)
    axes[1, 1].set_title('Error Distribution', fontsize=14, fontweight='bold')
    axes[1, 1].axvline(x=mae, color='red', linestyle='--', linewidth=2, label=f'MAE: {mae:.3f}')
    axes[1, 1].legend()
    axes[1, 1].grid(True, alpha=0.3)
    
    plt.tight_layout()
    
    # ÎπÑÍµê Ï†ÄÏû•
    comparison_path = output_path / 'comparison_with_gt.png'
    plt.savefig(comparison_path, dpi=300, bbox_inches='tight')
    print(f"\n   ‚úì ÎπÑÍµê ÏãúÍ∞ÅÌôî: {comparison_path}")
    
    plt.show()
    
    print("\n‚úÖ Ground Truth ÎπÑÍµê ÏôÑÎ£å!")
else:
    print(f"\n‚ö†Ô∏è  Ground TruthÎ•º Ï∞æÏùÑ Ïàò ÏóÜÏäµÎãàÎã§: {gt_path}")

# %% [markdown]
# ## 10. ÏôÑÎ£å!
#
# ÏòàÏ∏°Ïù¥ ÏôÑÎ£åÎêòÏóàÏäµÎãàÎã§! üéâ
#
# Í≤∞Í≥º ÌååÏùºÏùÄ `predictions/` ÎîîÎ†âÌÜ†Î¶¨Ïóê Ï†ÄÏû•ÎêòÏóàÏäµÎãàÎã§.

# %%
```

```python
# %% [markdown]
# # üîÆ Inverse Design Prediction (Sliding Window)
#
# ÌïôÏäµÎêú Inverse Design Î™®Îç∏Î°ú ÎåÄÌòï intensity map (4096√ó4096)ÏúºÎ°úÎ∂ÄÌÑ∞  
# pillar patternÏùÑ ÏÑ§Í≥ÑÌï©ÎãàÎã§.
#
# **Îç∞Ïù¥ÌÑ∞ ÌùêÎ¶Ñ:**
# - Input: Î™©Ìëú Phase Map (4096√ó4096 .npy)
# - Output: ÏÑ§Í≥ÑÎêú Pillar Pattern (4096√ó4096 PNG)
#
# ## üìã Î™©Ï∞®
# 1. ÌôòÍ≤Ω ÏÑ§Ï†ï Î∞è ÏûÑÌè¨Ìä∏
# 2. ÌååÎùºÎØ∏ÌÑ∞ ÏÑ§Ï†ï
# 3. Î™®Îç∏ Î°úÎìú
# 4. Ïä¨ÎùºÏù¥Îî© ÏúàÎèÑÏö∞ ÏòàÏ∏°
# 5. Ïù¥ÏßÑÌôî Î∞è Í≤∞Í≥º ÏãúÍ∞ÅÌôî

# %% [markdown]
# ## 1. ÌôòÍ≤Ω ÏÑ§Ï†ï Î∞è ÏûÑÌè¨Ìä∏

# %%
import torch
import numpy as np
import cv2
import matplotlib.pyplot as plt
from pathlib import Path
import sys
from tqdm import tqdm

# PyTorch ÏΩîÎìú Í≤ΩÎ°ú Ï∂îÍ∞Ä
sys.path.append('pytorch_codes')

from models import InverseUNet

# GPU ÌôïÏù∏
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(f"‚úÖ ÌôòÍ≤Ω ÏÑ§Ï†ï ÏôÑÎ£å!")
print(f"   Device: {device}")
print(f"   PyTorch Î≤ÑÏ†Ñ: {torch.__version__}")

# %% [markdown]
# ## 2. ÌååÎùºÎØ∏ÌÑ∞ ÏÑ§Ï†ï

# %%
# ==================== ÏûÖÎ†• ÌååÎùºÎØ∏ÌÑ∞ ====================
INPUT_PHASE_PATH = 'data/forward_intensity/outputs/sample_0000.npy'  # Î™©Ìëú intensity map
OUTPUT_DIR = 'predictions/inverse'                               # Ï∂úÎ†• ÎîîÎ†âÌÜ†Î¶¨

# ==================== Î™®Îç∏ ÌååÎùºÎØ∏ÌÑ∞ (ÌïôÏäµ ÏãúÏôÄ ÎèôÏùºÌï¥Ïïº Ìï®) ====================
LAYER_NUM = 5                       # U-Net Î†àÏù¥Ïñ¥ Ïàò
BASE_FEATURES = 64                  # Í∏∞Î≥∏ feature Ïàò
DROPOUT_RATE = 0.2                  # Dropout ÎπÑÏú®
USE_BATCHNORM = True                # BatchNorm ÏÇ¨Ïö© Ïó¨Î∂Ä
EXPERIMENT_NAME = 'inverse_design_basic_tiles'
CHECKPOINT_PATH = f'checkpoints/{EXPERIMENT_NAME}/best_model.pth'  # ÌïôÏäµÎêú Î™®Îç∏

# ==================== Ïä¨ÎùºÏù¥Îî© ÏúàÎèÑÏö∞ ÌååÎùºÎØ∏ÌÑ∞ ====================
TILE_SIZE = 256                     # ÌÉÄÏùº ÌÅ¨Í∏∞ (ÌïôÏäµ ÏãúÏôÄ ÎèôÏùºÌï¥Ïïº Ìï®)
STRIDE = 64                         # Ïä¨ÎùºÏù¥Îî© stride (ÏûëÏùÑÏàòÎ°ù Îçî Ï†ïÌôï, ÎäêÎ¶º)

# ==================== Ïù¥ÏßÑÌôî ÌååÎùºÎØ∏ÌÑ∞ ====================
THRESHOLD = 0.5                     # Pillar ÌôïÎ•† ÏûÑÍ≥ÑÍ∞í (ÎÖºÎ¨∏ Í∏∞Ï§Ä)

# Ï∂úÎ†• ÎîîÎ†âÌÜ†Î¶¨ ÏÉùÏÑ±
Path(OUTPUT_DIR).mkdir(parents=True, exist_ok=True)

print("‚úÖ ÌååÎùºÎØ∏ÌÑ∞ ÏÑ§Ï†ï ÏôÑÎ£å!")
print(f"\nüìä Inverse Design ÏÑ§Ï†ï:")
print(f"   ÏûÖÎ†• Phase Map: {INPUT_PHASE_PATH}")
print(f"   Ïã§Ìóò Ïù¥Î¶Ñ: {EXPERIMENT_NAME}")
print(f"   Ï≤¥ÌÅ¨Ìè¨Ïù∏Ìä∏: {CHECKPOINT_PATH}")
print(f"   ÌÉÄÏùº ÌÅ¨Í∏∞: {TILE_SIZE}√ó{TILE_SIZE}")
print(f"   Stride: {STRIDE}")
print(f"   Ïù¥ÏßÑÌôî ÏûÑÍ≥ÑÍ∞í: {THRESHOLD}")
print(f"   Device: {device}")

# %% [markdown]
# ## 3. ÏûÖÎ†• Phase Map Î°úÎìú

# %%
print("\nüìÇ ÏûÖÎ†• Phase Map Î°úÎî© Ï§ë...")

input_phase = np.load(INPUT_PHASE_PATH)

if input_phase is None:
    raise ValueError(f"Failed to load intensity map: {INPUT_PHASE_PATH}")

h, w = input_phase.shape

print(f"‚úÖ ÏûÖÎ†• Phase Map Î°úÎìú ÏôÑÎ£å!")
print(f"   ÌÅ¨Í∏∞: {w}√ó{h}")
print(f"   Phase Î≤îÏúÑ: [{input_phase.min():.2f}, {input_phase.max():.2f}]")

# ÏûÖÎ†• Phase Map ÏãúÍ∞ÅÌôî
plt.figure(figsize=(10, 10))
plt.imshow(input_phase, cmap='twilight')
plt.colorbar(label='Phase')
plt.title(f'Input Phase Map\n{w}√ó{h}\nRange: [{input_phase.min():.2f}, {input_phase.max():.2f}]')
plt.axis('off')
plt.tight_layout()
plt.show()

# %% [markdown]
# ## 4. Î™®Îç∏ Î°úÎìú

# %%
print("\nüì• Inverse Design Î™®Îç∏ Î°úÎî© Ï§ë...")

# Î™®Îç∏ ÏÉùÏÑ± (ÌïôÏäµ ÏãúÏôÄ ÎèôÏùºÌïú ÌååÎùºÎØ∏ÌÑ∞ ÏÇ¨Ïö©)
model = InverseUNet(
    in_channels=1,
    out_channels=[1],
    layer_num=LAYER_NUM,
    base_features=BASE_FEATURES,
    dropout_rate=DROPOUT_RATE,
    output_activations=['linear'],  # BCEWithLogitsLossÎ•º ÏúÑÌï¥ linear ÏÇ¨Ïö©
    use_batchnorm=USE_BATCHNORM
).to(device)

# Ï≤¥ÌÅ¨Ìè¨Ïù∏Ìä∏ Î°úÎìú
checkpoint = torch.load(CHECKPOINT_PATH, map_location=device)
model.load_state_dict(checkpoint['model_state_dict'])
model.eval()

print(f"‚úÖ Î™®Îç∏ Î°úÎìú ÏôÑÎ£å!")
print(f"   Epoch: {checkpoint.get('epoch', 'unknown')}")

# Val loss Ï∂úÎ†• (ÏûàÎäî Í≤ΩÏö∞ÏóêÎßå)
if 'val_loss' in checkpoint:
    print(f"   Val Loss: {checkpoint['val_loss']:.6f}")
    
# Î™®Îç∏ ÌååÎùºÎØ∏ÌÑ∞ Ïàò
total_params = sum(p.numel() for p in model.parameters())
print(f"   Total params: {total_params:,}")

# %% [markdown]
# ## 5. Phase Map Ï†ïÍ∑úÌôî

# %%
print("\nüîß Phase Map Ï†ïÍ∑úÌôî Ï§ë...")

# Phase mapÏùÑ [0, 1]Î°ú Ï†ïÍ∑úÌôî
phase_normalized = (input_phase - input_phase.min()) / (input_phase.max() - input_phase.min() + 1e-8)

print(f"‚úÖ Ï†ïÍ∑úÌôî ÏôÑÎ£å!")
print(f"   Ï†ïÍ∑úÌôî Î≤îÏúÑ: [{phase_normalized.min():.4f}, {phase_normalized.max():.4f}]")

# %% [markdown]
# ## 6. Ïä¨ÎùºÏù¥Îî© ÏúàÎèÑÏö∞ ÏòàÏ∏°

# %%
print("\nüîÆ Ïä¨ÎùºÏù¥Îî© ÏúàÎèÑÏö∞ ÏòàÏ∏° ÏãúÏûë...")

# ÏòàÏ∏° Í≤∞Í≥ºÎ•º ÎàÑÏ†ÅÌï† Î∞∞Ïó¥
prediction_sum = np.zeros((h, w), dtype=np.float32)
count_map = np.zeros((h, w), dtype=np.int32)

# Ïä¨ÎùºÏù¥Îî© ÏúàÎèÑÏö∞Î°ú ÌÉÄÏùº Ï∂îÏ∂ú Î∞è ÏòàÏ∏°
tiles_processed = 0
total_tiles = ((h - TILE_SIZE) // STRIDE + 1) * ((w - TILE_SIZE) // STRIDE + 1)

print(f"   Ï¥ù ÏòàÏ∏° ÌÉÄÏùº Ïàò: {total_tiles:,}")
print(f"   (Ïù¥ÎØ∏ÏßÄ ÌÅ¨Í∏∞ {h}√ó{w}, ÌÉÄÏùº {TILE_SIZE}√ó{TILE_SIZE}, stride {STRIDE})")

with torch.no_grad():
    # Y Î∞©Ìñ• Ïä¨ÎùºÏù¥Îî©
    for top in tqdm(range(0, h - TILE_SIZE + 1, STRIDE), desc="Y position"):
        # X Î∞©Ìñ• Ïä¨ÎùºÏù¥Îî©
        for left in range(0, w - TILE_SIZE + 1, STRIDE):
            # ÌÉÄÏùº Ï∂îÏ∂ú
            tile = phase_normalized[top:top+TILE_SIZE, left:left+TILE_SIZE]
            
            # TensorÎ°ú Î≥ÄÌôò [1, 1, H, W]
            tile_tensor = torch.from_numpy(tile).float().unsqueeze(0).unsqueeze(0).to(device)
            
            # ÏòàÏ∏° (logits)
            output = model(tile_tensor)
            
            # SigmoidÎ°ú ÌôïÎ•† Î≥ÄÌôò
            prob = torch.sigmoid(output).squeeze().cpu().numpy()
            
            # ÎàÑÏ†Å
            prediction_sum[top:top+TILE_SIZE, left:left+TILE_SIZE] += prob
            count_map[top:top+TILE_SIZE, left:left+TILE_SIZE] += 1
            
            tiles_processed += 1

print(f"\n‚úÖ ÏòàÏ∏° ÏôÑÎ£å! ({tiles_processed:,} ÌÉÄÏùº Ï≤òÎ¶¨)")

# %% [markdown]
# ## 7. Overlap Averaging

# %%
print("\nüìä Overlap Averaging Ï§ë...")

# ÌèâÍ∑† Í≥ÑÏÇ∞
average_prob_map = prediction_sum / np.maximum(count_map, 1)

print(f"‚úÖ Averaging ÏôÑÎ£å!")
print(f"   ÌôïÎ•† Î≤îÏúÑ: [{average_prob_map.min():.4f}, {average_prob_map.max():.4f}]")
print(f"   ÌèâÍ∑† ÌôïÎ•†: {average_prob_map.mean():.4f}")

# Count map ÌôïÏù∏
unique_counts = np.unique(count_map)
print(f"   Overlap ÌöüÏàò: {unique_counts.min()} ~ {unique_counts.max()}")

# %% [markdown]
# ## 8. Ïù¥ÏßÑÌôî (Binarization)

# %%
print(f"\nüî® Ïù¥ÏßÑÌôî Ï§ë (threshold={THRESHOLD})...")

# 0.5 ÏûÑÍ≥ÑÍ∞íÏúºÎ°ú Ïù¥ÏßÑÌôî (ÎÖºÎ¨∏ Î∞©Ïãù)
binary_pillar_pattern = (average_prob_map > THRESHOLD).astype(np.uint8) * 255

pillar_ratio = np.sum(binary_pillar_pattern > 0) / binary_pillar_pattern.size
print(f"‚úÖ Ïù¥ÏßÑÌôî ÏôÑÎ£å!")
print(f"   Pillar ÎπÑÏú®: {pillar_ratio * 100:.2f}%")

# %% [markdown]
# ## 9. Í≤∞Í≥º Ï†ÄÏû•

# %%
print("\nüíæ Í≤∞Í≥º Ï†ÄÏû• Ï§ë...")

# ÌååÏùº Ïù¥Î¶Ñ ÏÉùÏÑ±
input_name = Path(INPUT_PHASE_PATH).stem
output_prefix = f"{input_name}_inverse"

# 1. ÌôïÎ•† Îßµ Ï†ÄÏû• (.npy)
np.save(Path(OUTPUT_DIR) / f"{output_prefix}_prob_map.npy", average_prob_map)
print(f"   ‚úÖ ÌôïÎ•† Îßµ: {output_prefix}_prob_map.npy")

# 2. Ïù¥ÏßÑÌôîÎêú pillar pattern Ï†ÄÏû• (.png)
cv2.imwrite(str(Path(OUTPUT_DIR) / f"{output_prefix}_pillar_pattern.png"), binary_pillar_pattern)
print(f"   ‚úÖ Pillar Pattern: {output_prefix}_pillar_pattern.png")

# 3. Count map Ï†ÄÏû• (.npy)
np.save(Path(OUTPUT_DIR) / f"{output_prefix}_count_map.npy", count_map)
print(f"   ‚úÖ Count Map: {output_prefix}_count_map.npy")

print(f"\n‚úÖ Î™®Îì† Í≤∞Í≥º Ï†ÄÏû• ÏôÑÎ£å!")

# %% [markdown]
# ## 10. Í≤∞Í≥º ÏãúÍ∞ÅÌôî

# %%
print("\n" + "="*80)
print("üìä Í≤∞Í≥º ÏãúÍ∞ÅÌôî")
print("="*80)

fig, axes = plt.subplots(2, 3, figsize=(18, 12))

# 1. Input Phase Map
axes[0, 0].imshow(input_phase, cmap='twilight')
axes[0, 0].set_title(f'Input: Phase Map\nRange: [{input_phase.min():.2f}, {input_phase.max():.2f}]', fontsize=12)
axes[0, 0].axis('off')

# 2. Normalized Phase Map
axes[0, 1].imshow(phase_normalized, cmap='twilight')
axes[0, 1].set_title(f'Normalized Phase Map\nRange: [{phase_normalized.min():.4f}, {phase_normalized.max():.4f}]', fontsize=12)
axes[0, 1].axis('off')

# 3. Count Map
im3 = axes[0, 2].imshow(count_map, cmap='viridis')
axes[0, 2].set_title(f'Count Map (Overlap)\nRange: {unique_counts.min()} ~ {unique_counts.max()}', fontsize=12)
axes[0, 2].axis('off')
plt.colorbar(im3, ax=axes[0, 2], fraction=0.046)

# 4. Average Probability Map
im4 = axes[1, 0].imshow(average_prob_map, cmap='gray', vmin=0, vmax=1)
axes[1, 0].set_title(f'Probability Map\nMean: {average_prob_map.mean():.4f}', fontsize=12)
axes[1, 0].axis('off')
plt.colorbar(im4, ax=axes[1, 0], fraction=0.046)

# 5. Binary Pillar Pattern
axes[1, 1].imshow(binary_pillar_pattern, cmap='gray')
axes[1, 1].set_title(f'Binary Pillar Pattern (>{THRESHOLD})\nPillar: {pillar_ratio*100:.2f}%', fontsize=12)
axes[1, 1].axis('off')

# 6. Histogram
axes[1, 2].hist(average_prob_map.flatten(), bins=100, color='blue', alpha=0.7)
axes[1, 2].axvline(THRESHOLD, color='red', linestyle='--', linewidth=2, label=f'Threshold={THRESHOLD}')
axes[1, 2].set_xlabel('Probability', fontsize=11)
axes[1, 2].set_ylabel('Frequency', fontsize=11)
axes[1, 2].set_title('Probability Distribution', fontsize=12)
axes[1, 2].legend()
axes[1, 2].grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig(Path(OUTPUT_DIR) / f"{output_prefix}_visualization.png", dpi=150, bbox_inches='tight')
plt.show()

print(f"‚úÖ ÏãúÍ∞ÅÌôî Ï†ÄÏû•: {output_prefix}_visualization.png")

# %% [markdown]
# ## 11. ÏÉÅÏÑ∏ ÎπÑÍµê (Ï§åÏù∏)

# %%
print("\nüìä ÏÉÅÏÑ∏ ÏòÅÏó≠ ÎπÑÍµê (Ï§ëÏïô 512√ó512)")

# Ï§ëÏïô ÏòÅÏó≠ Ï∂îÏ∂ú
center_y, center_x = h // 2, w // 2
crop_size = 512
y1, y2 = center_y - crop_size//2, center_y + crop_size//2
x1, x2 = center_x - crop_size//2, center_x + crop_size//2

input_crop = input_phase[y1:y2, x1:x2]
prob_crop = average_prob_map[y1:y2, x1:x2]
binary_crop = binary_pillar_pattern[y1:y2, x1:x2]

# ÏãúÍ∞ÅÌôî
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

axes[0].imshow(input_crop, cmap='twilight')
axes[0].set_title(f'Input Phase Map\n(Center 512√ó512)', fontsize=14)
axes[0].axis('off')

axes[1].imshow(prob_crop, cmap='gray', vmin=0, vmax=1)
axes[1].set_title(f'Probability Map\n(Center 512√ó512)', fontsize=14)
axes[1].axis('off')

axes[2].imshow(binary_crop, cmap='gray')
axes[2].set_title(f'Binary Pillar Pattern\n(Center 512√ó512)', fontsize=14)
axes[2].axis('off')

plt.tight_layout()
plt.savefig(Path(OUTPUT_DIR) / f"{output_prefix}_zoom.png", dpi=150, bbox_inches='tight')
plt.show()

print(f"‚úÖ Ï§åÏù∏ Ïù¥ÎØ∏ÏßÄ Ï†ÄÏû•: {output_prefix}_zoom.png")

# %% [markdown]
# ## 12. ÌÜµÍ≥Ñ Ï†ïÎ≥¥

# %%
print("\n" + "="*80)
print("üìä Inverse Design Í≤∞Í≥º ÌÜµÍ≥Ñ")
print("="*80)

print(f"\n„ÄêInput Phase Map„Äë")
print(f"  ÌÅ¨Í∏∞: {input_phase.shape}")
print(f"  Î≤îÏúÑ: [{input_phase.min():.6f}, {input_phase.max():.6f}]")
print(f"  ÌèâÍ∑†: {input_phase.mean():.6f}")
print(f"  ÌëúÏ§ÄÌé∏Ï∞®: {input_phase.std():.6f}")

print(f"\n„ÄêPrediction Probability Map„Äë")
print(f"  ÌÅ¨Í∏∞: {average_prob_map.shape}")
print(f"  Î≤îÏúÑ: [{average_prob_map.min():.6f}, {average_prob_map.max():.6f}]")
print(f"  ÌèâÍ∑†: {average_prob_map.mean():.6f}")
print(f"  ÌëúÏ§ÄÌé∏Ï∞®: {average_prob_map.std():.6f}")

print(f"\n„ÄêBinary Pillar Pattern„Äë")
print(f"  ÌÅ¨Í∏∞: {binary_pillar_pattern.shape}")
print(f"  Pillar ÌîΩÏÖÄ Ïàò: {np.sum(binary_pillar_pattern > 0):,}")
print(f"  Pillar ÎπÑÏú®: {pillar_ratio * 100:.2f}%")
print(f"  Ïù¥ÏßÑÌôî ÏûÑÍ≥ÑÍ∞í: {THRESHOLD}")

print(f"\n„ÄêSliding Window„Äë")
print(f"  ÌÉÄÏùº ÌÅ¨Í∏∞: {TILE_SIZE}√ó{TILE_SIZE}")
print(f"  Stride: {STRIDE}")
print(f"  Ï≤òÎ¶¨Îêú ÌÉÄÏùº Ïàò: {tiles_processed:,}")
print(f"  Overlap Î≤îÏúÑ: {unique_counts.min()} ~ {unique_counts.max()}")

# %% [markdown]
# ## 13. ÏôÑÎ£å!
#
# Inverse DesignÏù¥ ÏôÑÎ£åÎêòÏóàÏäµÎãàÎã§! üéâ
#
# **ÏÉùÏÑ±Îêú ÌååÏùº:**
# - `*_prob_map.npy`: ÌèâÍ∑† ÌôïÎ•† Îßµ
# - `*_pillar_pattern.png`: Ïù¥ÏßÑÌôîÎêú pillar pattern (0.5 threshold)
# - `*_count_map.npy`: Overlap count map
# - `*_visualization.png`: Ï†ÑÏ≤¥ Í≤∞Í≥º ÏãúÍ∞ÅÌôî
# - `*_zoom.png`: Ï§ëÏïô ÏòÅÏó≠ ÌôïÎåÄ
#
# **Îã§Ïùå Îã®Í≥Ñ:**
# - ÏÑ§Í≥ÑÎêú pillar patternÏùÑ MEEPÏúºÎ°ú ÏãúÎÆ¨Î†àÏù¥ÏÖòÌïòÏó¨ Í≤ÄÏ¶ù
# - Îã§ÏñëÌïú intensity mapÏúºÎ°ú Ï∂îÍ∞Ä ÏÑ§Í≥Ñ ÌÖåÏä§Ìä∏

# %%
print("\n" + "="*80)
print("üéâ Inverse Design ÏôÑÎ£å!")
print("="*80)
print(f"\nüìÇ ÏÉùÏÑ±Îêú ÌååÏùº:")
print(f"   {Path(OUTPUT_DIR) / f'{output_prefix}_prob_map.npy'}")
print(f"   {Path(OUTPUT_DIR) / f'{output_prefix}_pillar_pattern.png'}")
print(f"   {Path(OUTPUT_DIR) / f'{output_prefix}_count_map.npy'}")
print(f"   {Path(OUTPUT_DIR) / f'{output_prefix}_visualization.png'}")
print(f"   {Path(OUTPUT_DIR) / f'{output_prefix}_zoom.png'}")
print(f"\nüí° ÏÑ§Í≥ÑÎêú pillar patternÏùÑ MEEPÏúºÎ°ú Í≤ÄÏ¶ùÌï¥Î≥¥ÏÑ∏Ïöî!")

```

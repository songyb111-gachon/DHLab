```python
"""
9가지 제작 오차 변형 비교 (2D 모니터, 즉시 시각화)
============================================================================

동일 마스크에 대해 9가지 (XY 픽셀 × angle) 조합 시뮬레이션 + 비교:

  XY 가장자리: -1, 0, +1 px (5nm)
  angle:       -1°, 0°, 1°
  높이:        350nm 고정

  → 3×3 = 9 cases
  → (0px, 0°) = 기준 (제작 오차 없음)

각 시뮬레이션 완료 시 즉시 기준 대비 비교 시각화.
2D 모니터 4개 사용 → 메모리 절약.

필요 패키지:
    pip install tidy3d numpy scipy matplotlib
"""

import os
import gc
import math
from typing import Dict, List

import numpy as np
from scipy import ndimage as ndi
from scipy.interpolate import RectBivariateSpline
import matplotlib.pyplot as plt

import tidy3d as td
from tidy3d import web
from tidy3d.plugins.polyslab import ComplexPolySlab

# ==================== 설정 ====================
RANDOM_SEED = 42
UNIT_PX = 128
SUPERCELL_NX = 4
SUPERCELL_NY = 4
NX = UNIT_PX * SUPERCELL_NX
NY = UNIT_PX * SUPERCELL_NY
PIXEL_SIZE_NM = 5.0
DZ_NM = 25.0
Z_TOTAL_NM = 3200

WAVELENGTH_NM = 1050
EPS_SIO2 = 1.4585
EPS_NANOPILLAR = 5.8
SUBSTRATE_THICKNESS_NM = 350
METASURFACE_HEIGHT_NM = 350

Z_SUBSTRATE_BOTTOM_NM = 1050
Z_SUBSTRATE_TOP_NM = Z_SUBSTRATE_BOTTOM_NM + SUBSTRATE_THICKNESS_NM
Z_METASURFACE_TOP_NM = Z_SUBSTRATE_TOP_NM + METASURFACE_HEIGHT_NM
Z_SOURCE_NM = 3050
PML_THICKNESS_NM = 150
PML_CELLS = int(PML_THICKNESS_NM / DZ_NM)
SIMULATION_TIME_FS = 500

Z_MONITOR_PLANES_NM = [
    Z_SUBSTRATE_TOP_NM,
    (Z_SUBSTRATE_TOP_NM + Z_METASURFACE_TOP_NM) / 2,
    Z_METASURFACE_TOP_NM,
    Z_METASURFACE_TOP_NM + 200,
]
Z_PLANE_LABELS = ["sub-meta", "meta-mid", "meta-top", "free-space"]

CTRL_GRID = 3
SPLINE_ORDER = 2
THRESHOLD = 0.0
BUFFER_PX = 8
SURFACE_GAUSSIAN_SIGMA = 0.5
MFS_PX = 5
MIN_ISLAND_AREA_PX = 13
MASK_EDGE_SMOOTH_SIGMA = 8.0

MASK_PIXEL_OFFSETS = [-1, 0, 1]
ANGLE_OFFSETS_DEG = [-1.0, 0.0, 1.0]

OUTPUT_DIR = "./9variant_comparison"
PRIMARY_PLANE = "meta-mid"
SECONDARY_PLANE = "meta-top"


class MaskGenerator:
    def __init__(self, seed=RANDOM_SEED):
        self.rng = np.random.default_rng(seed)

    def _make_quadrant_surface(self, ctrl, qn):
        g = ctrl.shape[0]
        gx, gy = np.linspace(0, 1, g), np.linspace(0, 1, g)
        k = max(1, min(SPLINE_ORDER, g - 1))
        spline = RectBivariateSpline(gx, gy, ctrl, kx=k, ky=k)
        S = spline(np.linspace(0, 1, qn), np.linspace(0, 1, qn))
        if SURFACE_GAUSSIAN_SIGMA > 0:
            S = ndi.gaussian_filter(S, sigma=SURFACE_GAUSSIAN_SIGMA, mode="reflect")
        return S

    def _fourfold_symmetry(self, quadrant):
        left = np.fliplr(quadrant)
        top = np.concatenate([left, quadrant], axis=1)
        bottom = np.flipud(top)
        return np.concatenate([bottom, top], axis=0)

    def _apply_buffer(self, mask):
        m = mask.astype(np.uint8).copy()
        b = BUFFER_PX
        m[:b, :] = m[-b:, :] = m[:, :b] = m[:, -b:] = 0
        return m

    def _mfs_filter(self, mask, mfs_px=MFS_PX):
        m = mask.astype(bool)
        if mfs_px > 1:
            r = max(1, int(np.round(mfs_px / 2)))
            se = ndi.iterate_structure(ndi.generate_binary_structure(2, 1), r)
            m = ndi.binary_closing(ndi.binary_opening(m, se), se)
        m = m.astype(np.uint8)
        labeled, n = ndi.label(m)
        if n > 0:
            counts = np.bincount(labeled.ravel())
            for lab in np.where(counts < MIN_ISLAND_AREA_PX)[0]:
                if lab != 0:
                    m[labeled == lab] = 0
        return m

    def _smooth_edges(self, mask):
        if MASK_EDGE_SMOOTH_SIGMA <= 0:
            return mask
        m = mask.astype(bool)
        sdf = ndi.distance_transform_edt(m) - ndi.distance_transform_edt(~m)
        sdf_s = ndi.gaussian_filter(sdf, sigma=MASK_EDGE_SMOOTH_SIGMA, mode="reflect")
        return (sdf_s > 0).astype(np.uint8)

    def generate_unit_cell(self, eps_value):
        ctrl = self.rng.uniform(-1, 1, size=(CTRL_GRID, CTRL_GRID))
        quad = self._make_quadrant_surface(ctrl, UNIT_PX // 2)
        field = self._fourfold_symmetry(quad)
        mask = (field > THRESHOLD).astype(np.uint8)
        mask = self._apply_buffer(mask)
        mask = self._mfs_filter(mask)
        mask = self._smooth_edges(mask)
        mask = self._mfs_filter(mask, mfs_px=1)
        return np.where(mask > 0, eps_value, 1.0).astype(np.float32)

    def generate_supercell(self):
        supercell = np.ones((NY, NX), dtype=np.float32)
        for iy in range(SUPERCELL_NY):
            for ix in range(SUPERCELL_NX):
                self.rng.uniform(0, 1)
                unit = self.generate_unit_cell(EPS_NANOPILLAR)
                y0, y1 = iy * UNIT_PX, (iy + 1) * UNIT_PX
                x0, x1 = ix * UNIT_PX, (ix + 1) * UNIT_PX
                supercell[y0:y1, x0:x1] = unit
        return supercell


# ============================================================================
# 마스크 오프셋 & 유틸
# ============================================================================

def nm_to_um(x): return x * 1e-3

DL_UM = nm_to_um(PIXEL_SIZE_NM)
DZ_UM = nm_to_um(DZ_NM)
DOMAIN_X_UM = NX * DL_UM
DOMAIN_Y_UM = NY * DL_UM
DOMAIN_Z_UM = nm_to_um(Z_TOTAL_NM)


def apply_mask_pixel_offset(eps_2d: np.ndarray, pixel_offset: int) -> np.ndarray:
    if pixel_offset == 0:
        return eps_2d.copy()
    binary = (eps_2d > 1.1).astype(np.uint8)
    se = ndi.generate_binary_structure(2, 1)
    if pixel_offset < 0:
        for _ in range(-pixel_offset):
            binary = ndi.binary_erosion(binary, structure=se)
    else:
        for _ in range(pixel_offset):
            binary = ndi.binary_dilation(binary, structure=se)
    return np.where(binary > 0, EPS_NANOPILLAR, 1.0).astype(np.float32)


def variant_label(pixel_offset: int, angle_deg: float) -> str:
    return f"px{pixel_offset:+d}_a{int(angle_deg):+d}"


def get_9_variants():
    variants = []
    for po in MASK_PIXEL_OFFSETS:
        for ad in ANGLE_OFFSETS_DEG:
            variants.append((po, ad))
    return variants


# ============================================================================
# 시뮬레이션 빌더 (2D 모니터)
# ============================================================================

def build_simulation(eps_2d: np.ndarray, angle_deg: float) -> td.Simulation:
    angle_rad = math.radians(angle_deg)
    Lx, Ly, Lz = DOMAIN_X_UM, DOMAIN_Y_UM, DOMAIN_Z_UM
    z_sub_bottom = nm_to_um(Z_SUBSTRATE_BOTTOM_NM)
    z_sub_top = nm_to_um(Z_SUBSTRATE_TOP_NM)
    z_meta_top = z_sub_top + nm_to_um(METASURFACE_HEIGHT_NM)
    z_source = nm_to_um(Z_SOURCE_NM)

    substrate = td.Structure(
        geometry=td.Box(
            center=(0, 0, (z_sub_bottom + z_sub_top) / 2),
            size=(td.inf, td.inf, z_sub_top - z_sub_bottom),
        ),
        medium=td.Medium(permittivity=EPS_SIO2),
        name="substrate",
    )
    structures: List[td.Structure] = [substrate]

    unique_eps = np.unique(eps_2d)
    for eps_val in unique_eps:
        if abs(eps_val - 1.0) < 0.01:
            continue
        mask = np.abs(eps_2d - eps_val) < 0.01
        if not mask.any():
            continue
        labeled, n_regions = ndi.label(mask)
        for region_id in range(1, n_regions + 1):
            region_mask = labeled == region_id
            ys, xs = np.where(region_mask)
            if len(xs) == 0:
                continue
            x_min, x_max = xs.min(), xs.max()
            y_min, y_max = ys.min(), ys.max()
            x0_um = (x_min - NX / 2) * DL_UM
            x1_um = (x_max - NX / 2 + 1) * DL_UM
            y0_um = (y_min - NY / 2) * DL_UM
            y1_um = (y_max - NY / 2 + 1) * DL_UM
            vertices = np.array(
                [[x0_um, y0_um], [x1_um, y0_um], [x1_um, y1_um], [x0_um, y1_um]]
            )
            slab_bounds = (z_sub_top, z_meta_top)

            if abs(angle_deg) < 1e-6:
                ps = td.PolySlab(vertices=vertices, slab_bounds=slab_bounds, axis=2)
                structures.append(
                    td.Structure(geometry=ps, medium=td.Medium(permittivity=float(eps_val)),
                                 name=f"pillar_{len(structures)}")
                )
            else:
                cps = ComplexPolySlab(
                    vertices=vertices, slab_bounds=slab_bounds, axis=2,
                    sidewall_angle=angle_rad, reference_plane="bottom",
                )
                mat = td.Medium(permittivity=float(eps_val))
                for sub in cps.sub_polyslabs:
                    structures.append(
                        td.Structure(geometry=sub, medium=mat, name=f"pillar_{len(structures)}")
                    )

    freq0 = td.C_0 / nm_to_um(WAVELENGTH_NM)
    fwidth = freq0 * 0.1
    source = td.PlaneWave(
        center=(0, 0, z_source), size=(td.inf, td.inf, 0),
        source_time=td.GaussianPulse(freq0=freq0, fwidth=fwidth),
        direction="-", pol_angle=0.0, name="source",
    )

    z_mon_min = nm_to_um(PML_THICKNESS_NM)
    z_mon_max = Lz - nm_to_um(PML_THICKNESS_NM)
    monitors = [
        td.FieldMonitor(
            center=(0, 0, (z_mon_min + z_mon_max) / 2),
            size=(Lx * 0.99, Ly * 0.99, z_mon_max - z_mon_min),
            freqs=[freq0], name="field_3d",
        )
    ]
    for z_nm, label in zip(Z_MONITOR_PLANES_NM, Z_PLANE_LABELS):
        monitors.append(
            td.FieldMonitor(
                center=(0, 0, nm_to_um(z_nm)),
                size=(Lx * 0.99, Ly * 0.99, 0),
                freqs=[freq0], name=f"field_{label}",
            )
        )

    z_diff_t = nm_to_um(Z_SUBSTRATE_BOTTOM_NM - 100)
    z_diff_r = nm_to_um(Z_SOURCE_NM - 100)
    monitors.append(
        td.DiffractionMonitor(
            center=(0, 0, z_diff_t), size=(td.inf, td.inf, 0),
            freqs=[freq0], name="diffraction_t", normal_dir="-",
        )
    )
    monitors.append(
        td.DiffractionMonitor(
            center=(0, 0, z_diff_r), size=(td.inf, td.inf, 0),
            freqs=[freq0], name="diffraction_r", normal_dir="+",
        )
    )

    boundary_spec = td.BoundarySpec(
        x=td.Boundary(plus=td.Periodic(), minus=td.Periodic()),
        y=td.Boundary(plus=td.Periodic(), minus=td.Periodic()),
        z=td.Boundary(plus=td.PML(num_layers=PML_CELLS), minus=td.PML(num_layers=PML_CELLS)),
    )

    grid_spec = td.GridSpec(
        grid_x=td.UniformGrid(dl=DL_UM),
        grid_y=td.UniformGrid(dl=DL_UM),
        grid_z=td.UniformGrid(dl=DZ_UM),
    )

    return td.Simulation(
        center=(0, 0, Lz / 2), size=(Lx, Ly, Lz),
        grid_spec=grid_spec, structures=structures, sources=[source],
        monitors=monitors, boundary_spec=boundary_spec,
        run_time=SIMULATION_TIME_FS * 1e-15,
    )


def extract_H_3d(sim_data, freq_idx: int = 0):
    """3D 모니터에서 H 필드 추출. 반환: H_real (Nx,Ny,Nz,3), H_imag (Nx,Ny,Nz,3) float32"""
    fd = sim_data["field_3d"]
    H_real_list, H_imag_list = [], []
    for comp in ["Hx", "Hy", "Hz"]:
        da = getattr(fd, comp)
        if "f" in da.dims:
            da = da.isel(f=freq_idx)
        arr = np.squeeze(da.values)
        H_real_list.append(np.real(arr))
        H_imag_list.append(np.imag(arr))
    H_real = np.stack(H_real_list, axis=-1).astype(np.float32)
    H_imag = np.stack(H_imag_list, axis=-1).astype(np.float32)
    return H_real, H_imag


def extract_H_2d(sim_data, monitor_name: str, freq_idx: int = 0) -> np.ndarray:
    fd = sim_data[monitor_name]
    components = []
    for comp in ["Hx", "Hy", "Hz"]:
        d = getattr(fd, comp)
        if "f" in d.dims:
            d = d.isel(f=freq_idx)
        components.append(np.squeeze(d.values))
    return np.stack(components, axis=-1)


def extract_diffraction(sim_data) -> Dict[str, np.ndarray]:
    """회절 모니터에서 차수별 파워 및 각도 추출"""
    result = {}
    for name in ("diffraction_t", "diffraction_r"):
        d = sim_data[name]
        power_raw = d.power
        power = np.squeeze(power_raw.values if hasattr(power_raw, "values") else np.asarray(power_raw))
        ox = d.orders_x
        oy = d.orders_y
        orders_x = np.asarray(ox.values if hasattr(ox, "values") else ox)
        orders_y = np.asarray(oy.values if hasattr(oy, "values") else oy)
        result[f"{name}_power"] = power
        result[f"{name}_orders_x"] = orders_x
        result[f"{name}_orders_y"] = orders_y
        result[f"{name}_total"] = float(power.sum())
    return result


def compute_errors(H_ref: np.ndarray, H_test: np.ndarray) -> Dict[str, float]:
    diff = np.abs(H_ref - H_test)
    denom_safe = np.where(np.abs(H_ref) > 1e-12, np.abs(H_ref), 1e-12)
    rel_rms = float(np.sqrt(np.mean((diff / denom_safe) ** 2)))
    l2_norm = float(np.linalg.norm(H_ref - H_test) / (np.linalg.norm(H_ref) + 1e-12))
    return {"relative_RMS": rel_rms, "L2_normalized": l2_norm}


# ============================================================================
# 시각화 — 회절 차수
# ============================================================================

def visualize_diffraction(diff_data, label_str, save_path):
    """단일 변형의 투과/반사 회절 효율 시각화"""
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))

    for ax, direction, title in zip(
        axes, ["diffraction_t", "diffraction_r"], ["Transmission", "Reflection"]
    ):
        power = diff_data[f"{direction}_power"]
        ox = diff_data[f"{direction}_orders_x"]
        oy = diff_data[f"{direction}_orders_y"]

        if power.ndim == 2 and power.shape[1] == 1:
            power_1d = power[:, 0]
            ax.bar(ox, power_1d, color="steelblue", edgecolor="k")
            ax.set_xlabel("Order (x)")
            ax.set_ylabel("Power")
        else:
            im = ax.imshow(
                power.T, origin="lower", cmap="YlOrRd", aspect="auto",
                extent=[ox[0] - 0.5, ox[-1] + 0.5, oy[0] - 0.5, oy[-1] + 0.5],
            )
            ax.set_xlabel("Order x")
            ax.set_ylabel("Order y")
            plt.colorbar(im, ax=ax, label="Power")
        total = diff_data[f"{direction}_total"]
        ax.set_title(f"{title} (total={total:.4f})")

    plt.suptitle(f"Diffraction — {label_str}", fontsize=13)
    plt.tight_layout()
    plt.savefig(save_path, dpi=150)
    plt.show()
    plt.close()


def visualize_diffraction_comparison(diff_ref, diff_test, label_str, save_path):
    """기준 vs 변형 회절 효율 비교 (투과 차수)"""
    fig, axes = plt.subplots(1, 3, figsize=(16, 5))

    ox_ref = diff_ref["diffraction_t_orders_x"]
    power_ref = diff_ref["diffraction_t_power"]
    power_test = diff_test["diffraction_t_power"]

    if power_ref.ndim == 2:
        power_ref = power_ref[:, power_ref.shape[1] // 2]
    if power_test.ndim == 2:
        power_test = power_test[:, power_test.shape[1] // 2]

    ox_test = diff_test["diffraction_t_orders_x"]
    common_min = max(ox_ref.min(), ox_test.min())
    common_max = min(ox_ref.max(), ox_test.max())
    mask_ref = (ox_ref >= common_min) & (ox_ref <= common_max)
    mask_test = (ox_test >= common_min) & (ox_test <= common_max)
    ox_c = ox_ref[mask_ref]
    pr = power_ref[mask_ref]
    pt = power_test[mask_test]

    w = 0.35
    x_pos = np.arange(len(ox_c))
    axes[0].bar(x_pos - w / 2, pr, w, label="ref", color="steelblue", edgecolor="k")
    axes[0].bar(x_pos + w / 2, pt, w, label=label_str, color="orange", edgecolor="k")
    axes[0].set_xticks(x_pos)
    axes[0].set_xticklabels([str(o) for o in ox_c], fontsize=8)
    axes[0].set_xlabel("Order x")
    axes[0].set_ylabel("Power")
    axes[0].set_title("Transmission orders")
    axes[0].legend(fontsize=8)

    diff_power = pt - pr
    colors = ["coral" if v > 0 else "steelblue" for v in diff_power]
    axes[1].bar(x_pos, diff_power, color=colors, edgecolor="k")
    axes[1].set_xticks(x_pos)
    axes[1].set_xticklabels([str(o) for o in ox_c], fontsize=8)
    axes[1].set_xlabel("Order x")
    axes[1].set_ylabel("ΔPower")
    axes[1].set_title("Power difference (test − ref)")
    axes[1].axhline(0, color="k", linewidth=0.5)

    axes[2].axis("off")
    table_rows = [["", "Ref", label_str, "Diff"]]
    table_rows.append([
        "T total",
        f"{diff_ref['diffraction_t_total']:.4f}",
        f"{diff_test['diffraction_t_total']:.4f}",
        f"{diff_test['diffraction_t_total'] - diff_ref['diffraction_t_total']:+.4f}",
    ])
    table_rows.append([
        "R total",
        f"{diff_ref['diffraction_r_total']:.4f}",
        f"{diff_test['diffraction_r_total']:.4f}",
        f"{diff_test['diffraction_r_total'] - diff_ref['diffraction_r_total']:+.4f}",
    ])
    t_sum = diff_ref["diffraction_t_total"] + diff_ref["diffraction_r_total"]
    t_sum2 = diff_test["diffraction_t_total"] + diff_test["diffraction_r_total"]
    table_rows.append(["T+R", f"{t_sum:.4f}", f"{t_sum2:.4f}", f"{t_sum2 - t_sum:+.4f}"])
    table = axes[2].table(cellText=table_rows, loc="center", cellLoc="center")
    table.auto_set_font_size(False)
    table.set_fontsize(10)
    table.scale(1, 1.5)
    for j in range(4):
        table[0, j].set_facecolor("#4472C4")
        table[0, j].set_text_props(color="white", fontweight="bold")
    axes[2].set_title("Power summary")

    plt.suptitle(f"Diffraction comparison — {label_str} vs ref", fontsize=13)
    plt.tight_layout()
    plt.savefig(save_path, dpi=150)
    plt.show()
    plt.close()


def visualize_diffraction_3x3(all_diff, variants, save_path):
    """3×3 회절 효율 비교 히트맵 (투과 0차 + 총 효율)"""
    ref_key = variant_label(0, 0.0)
    fig, axes = plt.subplots(1, 2, figsize=(14, 6))

    t0_grid = np.full((3, 3), np.nan)
    ttotal_grid = np.full((3, 3), np.nan)

    for idx, (po, ad) in enumerate(variants):
        key = variant_label(po, ad)
        row, col = idx // 3, idx % 3
        if key in all_diff:
            power = all_diff[key]["diffraction_t_power"]
            ox = all_diff[key]["diffraction_t_orders_x"]
            if power.ndim == 2:
                power = power[:, power.shape[1] // 2]
            zero_idx = np.argmin(np.abs(ox))
            t0_grid[row, col] = power[zero_idx]
            ttotal_grid[row, col] = all_diff[key]["diffraction_t_total"]

    for ax, grid, title in zip(
        axes, [t0_grid, ttotal_grid], ["0th order T", "Total T"]
    ):
        im = ax.imshow(grid, cmap="YlGnBu", origin="lower")
        for i in range(3):
            for j in range(3):
                val = grid[i, j]
                is_ref = (i == 1 and j == 1)
                txt = f"REF\n{val:.4f}" if is_ref else (f"{val:.4f}" if not np.isnan(val) else "—")
                ax.text(j, i, txt, ha="center", va="center", fontsize=9,
                        fontweight="bold" if is_ref else "normal")
        ax.set_xticks([0, 1, 2])
        ax.set_xticklabels([f"{a:+.0f}°" for a in ANGLE_OFFSETS_DEG])
        ax.set_yticks([0, 1, 2])
        ax.set_yticklabels([f"{p:+d}px" for p in MASK_PIXEL_OFFSETS])
        ax.set_xlabel("Sidewall angle")
        ax.set_ylabel("Pixel offset")
        ax.set_title(title)
        plt.colorbar(im, ax=ax, fraction=0.046)

    plt.suptitle("Diffraction efficiency — 9 variants", fontsize=14)
    plt.tight_layout()
    plt.savefig(save_path, dpi=150)
    plt.show()
    plt.close()


# ============================================================================
# 시각화 — 필드
# ============================================================================

def visualize_sample_3d(eps_2d, H_real, H_imag, label_str, save_path):
    """원본 fdtd_dataset_generator_tidy3d.py 동일 레이아웃 (3D 데이터 기반)
    Row 1: ε맵, |H|² xy-slice(z_mid), |H|² xz-slice(x_mid)
    Row 2: Re(Hy) xy, Im(Hy) xy, binary mask
    """
    H_mag = np.sum(H_real ** 2 + H_imag ** 2, axis=-1)
    nz = H_mag.shape[2]
    z_mid = nz // 2

    fig, axes = plt.subplots(2, 3, figsize=(15, 10))

    im = axes[0, 0].imshow(eps_2d, origin="lower", cmap="viridis")
    axes[0, 0].set_title(f"ε ({NX}×{NY})")
    plt.colorbar(im, ax=axes[0, 0])

    im = axes[0, 1].imshow(H_mag[:, :, z_mid].T, origin="lower", cmap="hot")
    axes[0, 1].set_title(f"|H|² (z={z_mid})")
    plt.colorbar(im, ax=axes[0, 1])

    im = axes[0, 2].imshow(
        H_mag[H_mag.shape[0] // 2, :, :].T, origin="lower", cmap="hot", aspect="auto"
    )
    axes[0, 2].set_title("|H|² (x-z slice)")
    plt.colorbar(im, ax=axes[0, 2])

    im = axes[1, 0].imshow(H_real[:, :, z_mid, 1].T, origin="lower", cmap="RdBu_r")
    axes[1, 0].set_title(f"Re(Hy) (z={z_mid})")
    plt.colorbar(im, ax=axes[1, 0])

    im = axes[1, 1].imshow(H_imag[:, :, z_mid, 1].T, origin="lower", cmap="RdBu_r")
    axes[1, 1].set_title(f"Im(Hy) (z={z_mid})")
    plt.colorbar(im, ax=axes[1, 1])

    binary = (eps_2d > 1.1).astype(float)
    im = axes[1, 2].imshow(binary, origin="lower", cmap="gray")
    axes[1, 2].set_title(f"Mask ({binary.mean() * 100:.1f}% fill)")
    plt.colorbar(im, ax=axes[1, 2])

    plt.suptitle(label_str, fontsize=14)
    plt.tight_layout()
    plt.savefig(save_path, dpi=150)
    plt.show()
    plt.close()


def visualize_comparison_3d(H_real_ref, H_imag_ref, H_real_test, H_imag_test,
                            label_str, save_path):
    """3D 기반 비교: 2×3 (test | ref | diff), xy-slice + xz-slice"""
    H_mag_ref = np.sum(H_real_ref ** 2 + H_imag_ref ** 2, axis=-1)
    H_mag_test = np.sum(H_real_test ** 2 + H_imag_test ** 2, axis=-1)
    nz = H_mag_ref.shape[2]
    z_mid = nz // 2
    x_mid = H_mag_ref.shape[0] // 2
    vmax_xy = max(H_mag_ref[:, :, z_mid].max(), H_mag_test[:, :, z_mid].max())
    vmax_xz = max(H_mag_ref[x_mid, :, :].max(), H_mag_test[x_mid, :, :].max())

    diff_xy = np.abs(H_mag_ref[:, :, z_mid] - H_mag_test[:, :, z_mid])
    diff_xz = np.abs(H_mag_ref[x_mid, :, :] - H_mag_test[x_mid, :, :])

    H_complex_ref = H_real_ref + 1j * H_imag_ref
    H_complex_test = H_real_test + 1j * H_imag_test
    l2_err = float(
        np.linalg.norm(H_complex_ref - H_complex_test)
        / (np.linalg.norm(H_complex_ref) + 1e-12)
    )

    fig, axes = plt.subplots(2, 3, figsize=(15, 10))

    im = axes[0, 0].imshow(
        H_mag_test[:, :, z_mid].T, origin="lower", cmap="hot", vmin=0, vmax=vmax_xy
    )
    axes[0, 0].set_title(f"|H|² {label_str} (z={z_mid})")
    plt.colorbar(im, ax=axes[0, 0])

    im = axes[0, 1].imshow(
        H_mag_ref[:, :, z_mid].T, origin="lower", cmap="hot", vmin=0, vmax=vmax_xy
    )
    axes[0, 1].set_title(f"|H|² ref (z={z_mid})")
    plt.colorbar(im, ax=axes[0, 1])

    im = axes[0, 2].imshow(diff_xy.T, origin="lower", cmap="magma")
    axes[0, 2].set_title(f"Δ|H|² (z={z_mid})")
    plt.colorbar(im, ax=axes[0, 2])

    im = axes[1, 0].imshow(
        H_mag_test[x_mid, :, :].T, origin="lower", cmap="hot",
        vmin=0, vmax=vmax_xz, aspect="auto",
    )
    axes[1, 0].set_title(f"|H|² {label_str} (x-z)")
    plt.colorbar(im, ax=axes[1, 0])

    im = axes[1, 1].imshow(
        H_mag_ref[x_mid, :, :].T, origin="lower", cmap="hot",
        vmin=0, vmax=vmax_xz, aspect="auto",
    )
    axes[1, 1].set_title("|H|² ref (x-z)")
    plt.colorbar(im, ax=axes[1, 1])

    im = axes[1, 2].imshow(diff_xz.T, origin="lower", cmap="magma", aspect="auto")
    axes[1, 2].set_title("Δ|H|² (x-z)")
    plt.colorbar(im, ax=axes[1, 2])

    plt.suptitle(
        f"{label_str} vs ref (3D)  —  L2={l2_err:.4f}",
        fontsize=13,
    )
    plt.tight_layout()
    plt.savefig(save_path, dpi=150)
    plt.show()
    plt.close()
    return l2_err


def visualize_fields(eps_2d, H_planes, label_str, save_path):
    """단일 변형 결과 — visualize_sample 2×3 레이아웃"""
    mn_pri = f"field_{PRIMARY_PLANE}"
    mn_sec = f"field_{SECONDARY_PLANE}"
    H_pri, H_sec = H_planes[mn_pri], H_planes[mn_sec]
    H_mag_pri = np.sum(np.abs(H_pri) ** 2, axis=-1)
    H_mag_sec = np.sum(np.abs(H_sec) ** 2, axis=-1)

    fig, axes = plt.subplots(2, 3, figsize=(15, 10))

    im = axes[0, 0].imshow(eps_2d, origin="lower", cmap="viridis")
    axes[0, 0].set_title(f"ε ({NX}×{NY})")
    plt.colorbar(im, ax=axes[0, 0])

    im = axes[0, 1].imshow(H_mag_pri.T, origin="lower", cmap="hot")
    axes[0, 1].set_title(f"|H|² @ {PRIMARY_PLANE}")
    plt.colorbar(im, ax=axes[0, 1])

    im = axes[0, 2].imshow(H_mag_sec.T, origin="lower", cmap="hot")
    axes[0, 2].set_title(f"|H|² @ {SECONDARY_PLANE}")
    plt.colorbar(im, ax=axes[0, 2])

    im = axes[1, 0].imshow(np.real(H_pri[:, :, 1]).T, origin="lower", cmap="RdBu_r")
    axes[1, 0].set_title(f"Re(Hy) @ {PRIMARY_PLANE}")
    plt.colorbar(im, ax=axes[1, 0])

    im = axes[1, 1].imshow(np.imag(H_pri[:, :, 1]).T, origin="lower", cmap="RdBu_r")
    axes[1, 1].set_title(f"Im(Hy) @ {PRIMARY_PLANE}")
    plt.colorbar(im, ax=axes[1, 1])

    binary = (eps_2d > 1.1).astype(float)
    im = axes[1, 2].imshow(binary, origin="lower", cmap="gray")
    axes[1, 2].set_title(f"Mask ({binary.mean() * 100:.1f}% fill)")
    plt.colorbar(im, ax=axes[1, 2])

    plt.suptitle(f"Fields — {label_str}", fontsize=14)
    plt.tight_layout()
    plt.savefig(save_path, dpi=150)
    plt.show()
    plt.close()


def visualize_comparison(H_ref, H_test, label_str, errors_at_plane, save_path):
    """변형 vs 기준 비교 — 2×3 (test | ref | diff)"""
    H_mag_ref = np.sum(np.abs(H_ref) ** 2, axis=-1)
    H_mag_test = np.sum(np.abs(H_test) ** 2, axis=-1)
    diff_mag = np.abs(H_mag_ref - H_mag_test)
    vmax_mag = max(H_mag_ref.max(), H_mag_test.max())

    Hy_ref_re = np.real(H_ref[:, :, 1])
    Hy_test_re = np.real(H_test[:, :, 1])
    diff_hy = np.abs(Hy_ref_re - Hy_test_re)
    vlim_hy = max(np.abs(Hy_ref_re).max(), np.abs(Hy_test_re).max())

    fig, axes = plt.subplots(2, 3, figsize=(15, 10))

    im = axes[0, 0].imshow(H_mag_test.T, origin="lower", cmap="hot", vmin=0, vmax=vmax_mag)
    axes[0, 0].set_title(f"|H|² {label_str}")
    plt.colorbar(im, ax=axes[0, 0])

    im = axes[0, 1].imshow(H_mag_ref.T, origin="lower", cmap="hot", vmin=0, vmax=vmax_mag)
    axes[0, 1].set_title("|H|² ref (0px, 0°)")
    plt.colorbar(im, ax=axes[0, 1])

    im = axes[0, 2].imshow(diff_mag.T, origin="lower", cmap="magma")
    axes[0, 2].set_title("Δ|H|²")
    plt.colorbar(im, ax=axes[0, 2])

    im = axes[1, 0].imshow(Hy_test_re.T, origin="lower", cmap="RdBu_r", vmin=-vlim_hy, vmax=vlim_hy)
    axes[1, 0].set_title(f"Re(Hy) {label_str}")
    plt.colorbar(im, ax=axes[1, 0])

    im = axes[1, 1].imshow(Hy_ref_re.T, origin="lower", cmap="RdBu_r", vmin=-vlim_hy, vmax=vlim_hy)
    axes[1, 1].set_title("Re(Hy) ref (0px, 0°)")
    plt.colorbar(im, ax=axes[1, 1])

    im = axes[1, 2].imshow(diff_hy.T, origin="lower", cmap="magma")
    axes[1, 2].set_title("ΔRe(Hy)")
    plt.colorbar(im, ax=axes[1, 2])

    plt.suptitle(
        f"{label_str} vs ref @ {PRIMARY_PLANE}  —  "
        f"L2={errors_at_plane['L2_normalized']:.4f}, "
        f"RMS={errors_at_plane['relative_RMS']:.4f}",
        fontsize=13,
    )
    plt.tight_layout()
    plt.savefig(save_path, dpi=150)
    plt.show()
    plt.close()


def visualize_3x3_grid(all_H, all_errors, variants, H_ref_pri, save_path):
    """3×3 그리드로 모든 변형의 |H|² 비교"""
    fig, axes = plt.subplots(3, 3, figsize=(15, 15))

    H_mag_ref = np.sum(np.abs(H_ref_pri) ** 2, axis=-1)
    global_vmax = H_mag_ref.max()
    for key in all_H:
        v = np.sum(np.abs(all_H[key]) ** 2, axis=-1).max()
        global_vmax = max(global_vmax, v)

    for idx, (po, ad) in enumerate(variants):
        row, col = idx // 3, idx % 3
        ax = axes[row, col]
        key = variant_label(po, ad)

        if key in all_H:
            H_mag = np.sum(np.abs(all_H[key]) ** 2, axis=-1)
            im = ax.imshow(H_mag.T, origin="lower", cmap="hot", vmin=0, vmax=global_vmax)
            if key in all_errors:
                e = all_errors[key]
                ax.set_title(f"px={po:+d}, {ad:+.0f}°\nL2={e['L2_normalized']:.4f}", fontsize=10)
            else:
                ax.set_title(f"px={po:+d}, {ad:+.0f}° (REF)", fontsize=10)
        else:
            ax.set_title(f"px={po:+d}, {ad:+.0f}° (pending)")
            ax.axis("off")

    plt.suptitle(f"|H|² @ {PRIMARY_PLANE} — 9 variants (vs 0px,0° ref)", fontsize=14)
    plt.tight_layout()
    plt.savefig(save_path, dpi=150)
    plt.show()
    plt.close()


def visualize_3x3_diff(all_H, all_errors, variants, H_ref_pri, save_path):
    """3×3 그리드로 모든 변형의 Δ|H|² 차이맵"""
    fig, axes = plt.subplots(3, 3, figsize=(15, 15))
    H_mag_ref = np.sum(np.abs(H_ref_pri) ** 2, axis=-1)
    ref_key = variant_label(0, 0.0)

    for idx, (po, ad) in enumerate(variants):
        row, col = idx // 3, idx % 3
        ax = axes[row, col]
        key = variant_label(po, ad)

        if key == ref_key:
            ax.text(0.5, 0.5, "REFERENCE\n(0px, 0°)", transform=ax.transAxes,
                    ha="center", va="center", fontsize=14, fontweight="bold")
            ax.set_facecolor("#f0f0f0")
            ax.set_xticks([])
            ax.set_yticks([])
            ax.set_title("px=0, 0° (REF)", fontsize=10)
        elif key in all_H:
            H_mag = np.sum(np.abs(all_H[key]) ** 2, axis=-1)
            diff = np.abs(H_mag - H_mag_ref)
            im = ax.imshow(diff.T, origin="lower", cmap="magma")
            plt.colorbar(im, ax=ax, fraction=0.046)
            e = all_errors[key]
            ax.set_title(f"px={po:+d}, {ad:+.0f}°\nL2={e['L2_normalized']:.4f}", fontsize=10)
        else:
            ax.set_title(f"px={po:+d}, {ad:+.0f}° (pending)")
            ax.axis("off")

    plt.suptitle(f"Δ|H|² vs reference @ {PRIMARY_PLANE}", fontsize=14)
    plt.tight_layout()
    plt.savefig(save_path, dpi=150)
    plt.show()
    plt.close()


def visualize_error_summary(all_errors, variants, save_path):
    """L2 오차 3×3 히트맵 + 바 차트"""
    fig, axes = plt.subplots(1, 2, figsize=(14, 6))

    l2_grid = np.full((3, 3), np.nan)
    rms_grid = np.full((3, 3), np.nan)
    ref_key = variant_label(0, 0.0)
    for idx, (po, ad) in enumerate(variants):
        key = variant_label(po, ad)
        row, col = idx // 3, idx % 3
        if key in all_errors:
            l2_grid[row, col] = all_errors[key]["L2_normalized"]
            rms_grid[row, col] = all_errors[key]["relative_RMS"]
        elif key == ref_key:
            l2_grid[row, col] = 0.0
            rms_grid[row, col] = 0.0

    ax = axes[0]
    im = ax.imshow(l2_grid, cmap="YlOrRd", origin="lower")
    for i in range(3):
        for j in range(3):
            val = l2_grid[i, j]
            txt = "REF" if (i == 1 and j == 1) else (f"{val:.4f}" if not np.isnan(val) else "—")
            ax.text(j, i, txt, ha="center", va="center", fontsize=10,
                    fontweight="bold" if i == 1 and j == 1 else "normal")
    ax.set_xticks([0, 1, 2])
    ax.set_xticklabels([f"{a:+.0f}°" for a in ANGLE_OFFSETS_DEG])
    ax.set_yticks([0, 1, 2])
    ax.set_yticklabels([f"{p:+d}px" for p in MASK_PIXEL_OFFSETS])
    ax.set_xlabel("Sidewall angle")
    ax.set_ylabel("Pixel offset")
    ax.set_title("L2 Error (vs ref)")
    plt.colorbar(im, ax=ax, fraction=0.046)

    ax = axes[1]
    keys_sorted = [variant_label(po, ad) for po, ad in variants if variant_label(po, ad) != ref_key]
    vals = []
    labels = []
    for po, ad in variants:
        key = variant_label(po, ad)
        if key == ref_key:
            continue
        if key in all_errors:
            vals.append(all_errors[key]["L2_normalized"])
            labels.append(f"{po:+d}px\n{ad:+.0f}°")
        else:
            vals.append(0)
            labels.append(f"{po:+d}px\n{ad:+.0f}°\n(—)")
    colors = ["steelblue"] * len(vals)
    max_idx = np.argmax(vals) if vals else 0
    colors[max_idx] = "coral"
    ax.bar(labels, vals, color=colors, edgecolor="k")
    ax.set_ylabel("L2 Error")
    ax.set_title("L2 Error by variant")
    if vals and max(vals) / (min(v for v in vals if v > 0) + 1e-30) > 5:
        ax.set_yscale("log")

    plt.suptitle(f"Error Summary @ {PRIMARY_PLANE}", fontsize=14)
    plt.tight_layout()
    plt.savefig(save_path, dpi=150)
    plt.show()
    plt.close()


# ============================================================================
# 메인
# ============================================================================

def main():
    os.makedirs(OUTPUT_DIR, exist_ok=True)
    mon_names = [f"field_{lb}" for lb in Z_PLANE_LABELS]
    mn_pri = f"field_{PRIMARY_PLANE}"
    variants = get_9_variants()
    ref_key = variant_label(0, 0.0)

    print("=" * 70)
    print("9가지 제작 오차 변형 비교 (2D + 3D 모니터)")
    print("=" * 70)
    print(f"  XY offset: {MASK_PIXEL_OFFSETS} px ({PIXEL_SIZE_NM}nm/px)")
    print(f"  Angle:     {ANGLE_OFFSETS_DEG}°")
    print(f"  Height:    {METASURFACE_HEIGHT_NM}nm (고정)")
    print(f"  DZ={DZ_NM}nm, time={SIMULATION_TIME_FS}fs")
    print(f"  모니터: 3D(field_3d) + 2D({Z_PLANE_LABELS})")
    print("=" * 70)

    mask_gen = MaskGenerator(seed=RANDOM_SEED)
    eps_2d_orig = mask_gen.generate_supercell()

    # ── 1단계: 기준 (0px, 0°) ──
    print(f"\n>>> [1/9] 기준: {ref_key} <<<")
    sim_ref = build_simulation(eps_2d_orig, 0.0)
    print(f"  Structures: {len(sim_ref.structures)}")
    sim_data_ref = web.run(sim_ref, task_name=f"9var_{ref_key}")

    H_ref_real, H_ref_imag = extract_H_3d(sim_data_ref)
    print(f"  3D H shape: {H_ref_real.shape}")

    ref_3d_path = os.path.join(OUTPUT_DIR, "_H_ref_3d.npz")
    np.savez(ref_3d_path, H_real=H_ref_real, H_imag=H_ref_imag)
    print(f"  3D ref 디스크 저장: {ref_3d_path}")

    H_ref_planes = {}
    for mn in mon_names:
        H_ref_planes[mn] = extract_H_2d(sim_data_ref, mn)

    diff_ref = extract_diffraction(sim_data_ref)
    print(f"  회절 T total={diff_ref['diffraction_t_total']:.4f}, "
          f"R total={diff_ref['diffraction_r_total']:.4f}")
    del sim_data_ref

    visualize_sample_3d(
        eps_2d_orig, H_ref_real, H_ref_imag, f"REF ({ref_key})",
        os.path.join(OUTPUT_DIR, "fields_ref_3d.png"),
    )
    del H_ref_real, H_ref_imag
    gc.collect()
    visualize_fields(
        eps_2d_orig, H_ref_planes, f"REF ({ref_key})",
        os.path.join(OUTPUT_DIR, "fields_ref_2d.png"),
    )
    visualize_diffraction(
        diff_ref, f"REF ({ref_key})",
        os.path.join(OUTPUT_DIR, "diffraction_ref.png"),
    )

    # ── 2단계: 나머지 8개 변형 ──
    all_H_pri = {ref_key: H_ref_planes[mn_pri]}
    all_errors = {}
    all_plane_errors = {}
    all_l2_3d = {}
    all_diff = {ref_key: diff_ref}

    step = 1
    for po, ad in variants:
        key = variant_label(po, ad)
        if key == ref_key:
            continue
        step += 1
        print(f"\n>>> [{step}/9] {key} (px={po:+d}, angle={ad:+.0f}°) <<<")

        eps_2d_var = apply_mask_pixel_offset(eps_2d_orig, po)
        sim = build_simulation(eps_2d_var, ad)
        print(f"  Structures: {len(sim.structures)}")
        sim_data = web.run(sim, task_name=f"9var_{key}")

        H_test_real, H_test_imag = extract_H_3d(sim_data)
        print(f"  3D H shape: {H_test_real.shape}")

        H_test_planes = {}
        plane_errs = {}
        for mn, lb in zip(mon_names, Z_PLANE_LABELS):
            H_test = extract_H_2d(sim_data, mn)
            H_test_planes[mn] = H_test
            plane_errs[lb] = compute_errors(H_ref_planes[mn], H_test)
            print(f"    {lb}: L2={plane_errs[lb]['L2_normalized']:.4f}")

        diff_test = extract_diffraction(sim_data)
        all_diff[key] = diff_test
        print(f"    회절 T={diff_test['diffraction_t_total']:.4f}, "
              f"R={diff_test['diffraction_r_total']:.4f}")
        del sim_data

        all_H_pri[key] = H_test_planes[mn_pri]
        all_errors[key] = plane_errs[PRIMARY_PLANE]
        all_plane_errors[key] = plane_errs

        visualize_sample_3d(
            eps_2d_var, H_test_real, H_test_imag, key,
            os.path.join(OUTPUT_DIR, f"fields_3d_{key}.png"),
        )

        ref_3d = np.load(ref_3d_path)
        l2_3d = visualize_comparison_3d(
            ref_3d["H_real"], ref_3d["H_imag"], H_test_real, H_test_imag, key,
            os.path.join(OUTPUT_DIR, f"compare_3d_{key}.png"),
        )
        del ref_3d
        all_l2_3d[key] = l2_3d
        print(f"  3D L2={l2_3d:.4f}")

        del H_test_real, H_test_imag
        gc.collect()

        visualize_comparison(
            H_ref_planes[mn_pri], H_test_planes[mn_pri],
            key, plane_errs[PRIMARY_PLANE],
            os.path.join(OUTPUT_DIR, f"compare_2d_{key}.png"),
        )

        visualize_diffraction_comparison(
            diff_ref, diff_test, key,
            os.path.join(OUTPUT_DIR, f"diffraction_{key}.png"),
        )

        visualize_3x3_grid(
            all_H_pri, all_errors, variants, H_ref_planes[mn_pri],
            os.path.join(OUTPUT_DIR, f"grid_step{step}.png"),
        )

    # ── 최종 시각화 ──
    visualize_3x3_diff(
        all_H_pri, all_errors, variants, H_ref_planes[mn_pri],
        os.path.join(OUTPUT_DIR, "final_diff_3x3.png"),
    )

    visualize_error_summary(
        all_errors, variants,
        os.path.join(OUTPUT_DIR, "final_error_summary.png"),
    )

    visualize_diffraction_3x3(
        all_diff, variants,
        os.path.join(OUTPUT_DIR, "final_diffraction_3x3.png"),
    )

    # ── 테이블 출력 ──
    print("\n" + "=" * 70)
    print(f"{'Variant':>12}  {'3D L2':>10}  {'2D L2':>10}  {'2D RMS':>10}  ", end="")
    for lb in Z_PLANE_LABELS:
        print(f" {lb+' L2':>12}", end="")
    print()
    print("-" * (46 + 13 * len(Z_PLANE_LABELS)))
    for po, ad in variants:
        key = variant_label(po, ad)
        if key == ref_key:
            print(f"{key:>12}  {'—(ref)':>10}  {'—(ref)':>10}  {'—(ref)':>10}  ", end="")
            for lb in Z_PLANE_LABELS:
                print(f" {'0.00000':>12}", end="")
            print()
        elif key in all_plane_errors:
            e_pri = all_errors[key]
            l3d = all_l2_3d.get(key, 0.0)
            print(
                f"{key:>12}  {l3d:>10.5f}  {e_pri['L2_normalized']:>10.5f}  "
                f"{e_pri['relative_RMS']:>10.5f}  ", end=""
            )
            for lb in Z_PLANE_LABELS:
                print(f" {all_plane_errors[key][lb]['L2_normalized']:>12.5f}", end="")
            print()
    print("=" * 70)

    # ── 저장 ──
    save_dict = {
        "eps_2d_orig": eps_2d_orig,
        "plane_labels": Z_PLANE_LABELS,
        "pixel_offsets": np.array(MASK_PIXEL_OFFSETS),
        "angle_offsets": np.array(ANGLE_OFFSETS_DEG),
    }
    for mn, lb in zip(mon_names, Z_PLANE_LABELS):
        save_dict[f"H_ref_{lb}"] = H_ref_planes[mn]
    for key in all_H_pri:
        save_dict[f"H_pri_{key}"] = all_H_pri[key]
    for key in all_errors:
        save_dict[f"l2_2d_{key}"] = all_errors[key]["L2_normalized"]
        save_dict[f"rms_2d_{key}"] = all_errors[key]["relative_RMS"]
    for key in all_l2_3d:
        save_dict[f"l2_3d_{key}"] = all_l2_3d[key]
    for key in all_diff:
        save_dict[f"diff_t_total_{key}"] = all_diff[key]["diffraction_t_total"]
        save_dict[f"diff_r_total_{key}"] = all_diff[key]["diffraction_r_total"]
        save_dict[f"diff_t_power_{key}"] = all_diff[key]["diffraction_t_power"]
        save_dict[f"diff_t_orders_x_{key}"] = all_diff[key]["diffraction_t_orders_x"]

    np.savez(os.path.join(OUTPUT_DIR, "9variant_results.npz"), **save_dict)
    print(f"\n결과 저장: {OUTPUT_DIR}/9variant_results.npz")
    print("완료!")


if __name__ == "__main__":
    main()
```

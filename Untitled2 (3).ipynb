```python
"""
BeamSystem 시각화 테스트
- 시스템 수차 적용 확인
- 보정 효과 확인
- PSNR 변화 확인
"""

import torch
import sys
import os
import numpy as np
import matplotlib.pyplot as plt

try:
    base_dir = os.path.dirname(__file__)
except NameError:
    base_dir = os.getcwd()

sys.path.append(os.path.abspath(os.path.join(base_dir, '..')))

from zoo.beam_system import BeamSystem
from core.config import ExperimentConfig


# ============================================================
#  로컬 오버라이드  (주석 해제 후 값을 수정하면 config.py 기본값 대신 적용)
# ============================================================
OVERRIDES = {
    # ── 광학 ──
    # "physics.wavelength":           638e-9,
    # "physics.pixel_size":           6.3e-6,
    # "physics.grid_size":            256,
    # "physics.max_zernike_modes":    15,
    # "physics.beam_waist":           1e-3,
    # "physics.aberration_scale":     1.0,
    # "physics.n_dust_particles":    5,
    # "physics.dust_scale":          1.0,
    # "physics.laser_to_slm_distance":  5e-3,
    # "physics.slm_to_camera_distance": 5e-3,
}


def test_visualization():
    """BeamSystem 시각화 테스트"""
    config = ExperimentConfig()
    if OVERRIDES:
        config.update(OVERRIDES)

    print("=" * 60)
    print("BeamSystem Visualization Test")
    print("=" * 60)
    print(f"Device: {config.device}")

    system = BeamSystem(
        grid_size=config.physics.grid_size,
        pixel_size=config.physics.pixel_size,
        wavelength=config.physics.wavelength,
        laser_to_slm_distance=config.physics.laser_to_slm_distance,
        slm_to_camera_distance=config.physics.slm_to_camera_distance,
        max_zernike_modes=config.physics.max_zernike_modes,
        beam_waist=config.physics.beam_waist,
        aberration_scale=config.physics.aberration_scale,
        n_dust_particles=config.physics.n_dust_particles,
        dust_scale=config.physics.dust_scale,
        device=config.device,
    )

    print(f"\nGrid size: {system.grid_size}")
    print(f"Zernike modes: {system.max_zernike_modes}")
    print(f"Aberration scale: {system.aberration_scale}")
    print(f"Dust particles: {system.n_dust_particles}")

    # === 1. 이상적 PSF vs 수차 PSF 비교 ===
    fig, axes = plt.subplots(2, 4, figsize=(16, 8))
    fig.suptitle("BeamSystem Simulation Test", fontsize=14)

    ideal_psf = system.ideal_psf.cpu().numpy()
    axes[0, 0].imshow(ideal_psf, cmap="hot")
    axes[0, 0].set_title("Ideal PSF (No Aberration)")
    axes[0, 0].axis("off")

    batch_size = 1
    focus, _ = system.reset(batch_size=batch_size)
    aberrated_np = focus[0].cpu().numpy()

    axes[0, 1].imshow(aberrated_np, cmap="hot")
    axes[0, 1].set_title("Aberrated PSF (Random System Aberration)")
    axes[0, 1].axis("off")

    diff_before = np.abs(aberrated_np - ideal_psf)
    axes[0, 2].imshow(diff_before, cmap="viridis")
    axes[0, 2].set_title("Difference (Before Correction)")
    axes[0, 2].axis("off")

    coeffs = system.zernike_coefficients[0].cpu().numpy()
    zernike_names = [
        "Piston", "Tilt X", "Tilt Y", "Defocus", "Astig 45", "Astig 0",
        "Coma Y", "Coma X", "Trefoil Y", "Trefoil X", "Spherical",
        "Astig2 45", "Astig2 0", "Quadfoil Y", "Quadfoil X",
    ]
    colors = ["green" if c >= 0 else "red" for c in coeffs]
    axes[0, 3].barh(range(len(coeffs)), coeffs, color=colors, alpha=0.7)
    axes[0, 3].set_yticks(range(len(coeffs)))
    axes[0, 3].set_yticklabels(zernike_names[: len(coeffs)], fontsize=8)
    axes[0, 3].set_xlabel("Coefficient Value")
    axes[0, 3].set_title("System Aberration Coefficients")
    axes[0, 3].axvline(x=0, color="black", linestyle="-", linewidth=0.5)

    metrics_before = system.compute_metrics(focus)
    print(f"\n=== Before Correction ===")
    print(f"PSNR: {metrics_before['psnr']:.2f} dB")

    # === 2. 최적 보정 적용 ===
    optimal_correction = system.get_optimal_correction()
    corrected_focus, _ = system.forward(optimal_correction)
    corrected_np = corrected_focus[0].cpu().numpy()

    axes[1, 0].imshow(ideal_psf, cmap="hot")
    axes[1, 0].set_title("Ideal PSF (Reference)")
    axes[1, 0].axis("off")

    axes[1, 1].imshow(corrected_np, cmap="hot")
    axes[1, 1].set_title("Corrected PSF (After Optimal Correction)")
    axes[1, 1].axis("off")

    diff_after = np.abs(corrected_np - ideal_psf)
    axes[1, 2].imshow(diff_after, cmap="viridis")
    axes[1, 2].set_title("Difference (After Correction)")
    axes[1, 2].axis("off")

    corr_coeffs = optimal_correction[0].cpu().numpy()
    n_z = system._dim_zernike
    corr_labels = zernike_names[1 : n_z + 1]
    if system._dim_dust > 0:
        for d in range(system.n_dust_particles):
            for p in ["x", "y", "z", "a"]:
                corr_labels.append(f"D{d}_{p}")
    colors = ["green" if c >= 0 else "red" for c in corr_coeffs]
    axes[1, 3].barh(range(len(corr_coeffs)), corr_coeffs, color=colors, alpha=0.7)
    axes[1, 3].set_yticks(range(len(corr_coeffs)))
    axes[1, 3].set_yticklabels(corr_labels, fontsize=6)
    axes[1, 3].set_xlabel("Coefficient Value")
    axes[1, 3].set_title("Optimal Correction")
    axes[1, 3].axvline(x=0, color="black", linestyle="-", linewidth=0.5)

    metrics_after = system.compute_metrics(corrected_focus)
    print(f"\n=== After Optimal Correction ===")
    print(f"PSNR: {metrics_after['psnr']:.2f} dB")
    print(f"PSNR Improvement: {metrics_after['psnr'] - metrics_before['psnr']:.2f} dB")

    plt.tight_layout()
    plt.savefig("beam_system_test.png", dpi=150)
    plt.show()

    # === 3. 5가지 시나리오 비교 ===
    # (수차 종류) × (보정 모드) 조합을 각각 독립 실행
    #
    #   시나리오 1: Zernike 수차만 → Zernike 보정
    #   시나리오 2: Dust 수차만   → Dust 보정
    #   시나리오 3: 둘 다 수차    → Zernike만 보정
    #   시나리오 4: 둘 다 수차    → Dust만 보정
    #   시나리오 5: 둘 다 수차    → 둘 다 보정
    n_steps = 10
    dim_z = system._dim_zernike
    dim_d = system._dim_dust

    print("\n" + "=" * 60)
    print(f"5-Scenario Comparison ({n_steps} steps each)")
    print("=" * 60)

    # 기준 수차 생성 (모든 시나리오에서 동일한 값 사용)
    system.reset(batch_size=1)
    ref_zernike = system.zernike_coefficients.clone()         # (1, modes)
    ref_dust = system.initial_dust_params.clone() if dim_d > 0 else None
    zero_zernike = torch.zeros_like(ref_zernike)
    zero_zernike[:, 0] = 0.0  # piston은 항상 0
    zero_dust = torch.zeros_like(ref_dust) if ref_dust is not None else None

    scenarios = [
        # (이름,            사용할 Zernike,   사용할 Dust, Zernike 보정, Dust 보정)
        ("1. Zernike Aberr → Z Corr",  ref_zernike,  zero_dust,   True,  False),
        ("2. Dust Aberr → D Corr",     zero_zernike, ref_dust,    False, True),
        ("3. Both Aberr → Z Only",     ref_zernike,  ref_dust,    True,  False),
        ("4. Both Aberr → D Only",     ref_zernike,  ref_dust,    False, True),
        ("5. Both Aberr → Both Corr",  ref_zernike,  ref_dust,    True,  True),
    ]

    all_results = {}

    for sc_name, sc_zernike, sc_dust, corr_z, corr_d in scenarios:
        system.reset_from_coefficients(sc_zernike, dust=sc_dust)
        focus, _ = system._forward_pass(system.zernike_coefficients)

        optimal = system.get_optimal_correction()
        masked = torch.zeros_like(optimal)
        if corr_z:
            masked[:, :dim_z] = optimal[:, :dim_z]
        if corr_d and dim_d > 0:
            masked[:, dim_z:] = optimal[:, dim_z:]
        step_action = masked / n_steps

        psnr_list = [system.compute_metrics(focus)["psnr"]]
        rmse_list = [torch.norm(system.zernike_coefficients[:, 1:], dim=1).item()]
        psf_list = [focus[0].cpu().numpy()]

        for s in range(n_steps):
            focus, _ = system.forward(step_action)
            psnr_list.append(system.compute_metrics(focus)["psnr"])
            rmse_list.append(torch.norm(system.zernike_coefficients[:, 1:], dim=1).item())
            psf_list.append(focus[0].cpu().numpy())

        all_results[sc_name] = {
            "psnr": psnr_list, "rmse": rmse_list, "psf": psf_list
        }
        print(f"\n  [{sc_name}]")
        print(f"    PSNR: {psnr_list[0]:.2f} → {psnr_list[-1]:.2f} dB "
              f"(+{psnr_list[-1] - psnr_list[0]:.2f})")
        print(f"    RMSE: {rmse_list[0]:.4f} → {rmse_list[-1]:.4f}")

    # ---- 시나리오별 개별 시각화 (PSF + PSNR/RMSE 그래프) ----
    show_idx = [0, 2, 5, 8, 10]
    for sc_name, data in all_results.items():
        safe_name = sc_name.replace(" ", "_").replace("→", "to")

        # PSF 이미지
        fig_sc, axes_sc = plt.subplots(2, len(show_idx),
                                       figsize=(4 * len(show_idx), 7))
        fig_sc.suptitle(sc_name, fontsize=14, fontweight='bold')

        for col, si in enumerate(show_idx):
            axes_sc[0, col].imshow(data["psf"][si], cmap="hot")
            axes_sc[0, col].set_title(
                f"Step {si}\nPSNR: {data['psnr'][si]:.1f} dB", fontsize=9)
            axes_sc[0, col].axis("off")

            diff = np.abs(data["psf"][si] - ideal_psf)
            axes_sc[1, col].imshow(diff, cmap="viridis")
            axes_sc[1, col].set_title("Diff from Ideal", fontsize=9)
            axes_sc[1, col].axis("off")

        plt.tight_layout()
        plt.savefig(f"beam_scenario_{safe_name}_psf.png", dpi=150)
        plt.show()

        # PSNR + RMSE 그래프
        fig_m, (ax_mp, ax_mr) = plt.subplots(1, 2, figsize=(14, 5))
        fig_m.suptitle(f"{sc_name} — Metrics", fontsize=13, fontweight='bold')

        ax_mp.plot(range(n_steps + 1), data["psnr"], "bo-",
                   linewidth=2, markersize=7)
        ax_mp.set_xlabel("Step", fontsize=12)
        ax_mp.set_ylabel("PSNR (dB)", fontsize=12)
        ax_mp.set_title("PSNR vs Step")
        ax_mp.set_xticks(range(n_steps + 1))
        ax_mp.grid(True, alpha=0.3)

        ax_mr.plot(range(n_steps + 1), data["rmse"], "rs-",
                   linewidth=2, markersize=7)
        ax_mr.set_xlabel("Step", fontsize=12)
        ax_mr.set_ylabel("Zernike RMSE (L2 norm)", fontsize=12)
        ax_mr.set_title("RMSE vs Step")
        ax_mr.set_xticks(range(n_steps + 1))
        ax_mr.grid(True, alpha=0.3)

        plt.tight_layout()
        plt.savefig(f"beam_scenario_{safe_name}_metrics.png", dpi=150)
        plt.show()

    # ---- 전체 비교 PSNR + RMSE 그래프 ----
    sc_colors = ["blue", "red", "orange", "purple", "green"]
    sc_markers = ["o", "s", "^", "D", "P"]

    fig_cmp, (ax_p, ax_r) = plt.subplots(1, 2, figsize=(18, 7))
    fig_cmp.suptitle("5-Scenario Comparison", fontsize=14, fontweight='bold')

    for i, (sc_name, data) in enumerate(all_results.items()):
        ax_p.plot(range(n_steps + 1), data["psnr"],
                  marker=sc_markers[i], color=sc_colors[i],
                  linewidth=2, markersize=6, label=sc_name)
        ax_r.plot(range(n_steps + 1), data["rmse"],
                  marker=sc_markers[i], color=sc_colors[i],
                  linewidth=2, markersize=6, label=sc_name)

    ax_p.set_xlabel("Step", fontsize=12)
    ax_p.set_ylabel("PSNR (dB)", fontsize=12)
    ax_p.set_title("PSNR vs Correction Step")
    ax_p.set_xticks(range(n_steps + 1))
    ax_p.legend(fontsize=8, loc="best")
    ax_p.grid(True, alpha=0.3)

    ax_r.set_xlabel("Step", fontsize=12)
    ax_r.set_ylabel("Zernike RMSE (L2 norm)", fontsize=12)
    ax_r.set_title("Zernike RMSE vs Correction Step")
    ax_r.set_xticks(range(n_steps + 1))
    ax_r.legend(fontsize=8, loc="best")
    ax_r.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.savefig("beam_scenario_comparison.png", dpi=150)
    plt.show()

    # ---- 최종 결과 요약 표 ----
    print("\n" + "=" * 60)
    print("Summary")
    print("=" * 60)
    print(f"{'Scenario':<35} {'PSNR_i':>7} {'PSNR_f':>7} {'dPSNR':>7} "
          f"{'RMSE_i':>8} {'RMSE_f':>8}")
    print("-" * 80)
    for sc_name, data in all_results.items():
        dp = data['psnr'][-1] - data['psnr'][0]
        print(f"{sc_name:<35} {data['psnr'][0]:7.2f} {data['psnr'][-1]:7.2f} "
              f"{dp:+7.2f} {data['rmse'][0]:8.4f} {data['rmse'][-1]:8.4f}")
    print("=" * 60)


if __name__ == "__main__":
    test_visualization()
```

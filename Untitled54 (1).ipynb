```python
# ÏàúÏàò ÏòµÌã∞ÎßàÏù¥Ï†ÄÎßåÏúºÎ°ú ÏàòÏ∞® Î≥¥Ï†ï Î¨∏Ï†ú Ìï¥Í≤∞ Í∞ÄÎä•ÌïúÏßÄ ÌÖåÏä§Ìä∏
# Gradient-free optimizationÏúºÎ°ú Zernike Í≥ÑÏàòÎ•º ÏßÅÏ†ë ÏµúÏ†ÅÌôî

import torch
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize
from config import config
import sys

sys.path.append('./torchOptics')

# ÏãúÎÆ¨Î†àÏù¥ÌÑ∞ ÏÑ†ÌÉù
SIMULATOR_TYPE = config.DEFAULT_SIMULATOR_TYPE  # 'point_source' ÎòêÎäî 'beam'

if SIMULATOR_TYPE == 'point_source':
    from point_source_simulator import PointSourceSimulator
    print("‚úÖ Ï†êÍ¥ëÏõê ÏãúÎÆ¨Î†àÏù¥ÌÑ∞ ÏÇ¨Ïö©")
else:
    from beam_simulator import BeamSimulator
    print("‚úÖ Îπî ÏãúÎÆ¨Î†àÏù¥ÌÑ∞ ÏÇ¨Ïö©")

def test_optimizer_correction(
    optimizer_name='NelderMead',
    learning_rate=0.1,
    max_iterations=200,
    convergence_threshold=0.99,
    image_size=16,
    max_zernike_order=15
):
    """
    ÏàúÏàò ÏòµÌã∞ÎßàÏù¥Ï†ÄÎßåÏúºÎ°ú ÏàòÏ∞® Î≥¥Ï†ï ÏãúÎèÑ
    
    Args:
        optimizer_name: ÏòµÌã∞ÎßàÏù¥Ï†Ä Ï¢ÖÎ•ò ('NelderMead', 'Powell', 'COBYLA', 'BFGS')
        learning_rate: ÌïôÏäµÎ•† (scipy ÏòµÌã∞ÎßàÏù¥Ï†ÄÏóêÏÑúÎäî ÏÇ¨Ïö© Ïïà Ìï®)
        max_iterations: ÏµúÎåÄ Î∞òÎ≥µ ÌöüÏàò
        convergence_threshold: ÏàòÎ†¥ ÌåêÏ†ï Strehl ratio ÏûÑÍ≥ÑÍ∞í
        image_size: Ïù¥ÎØ∏ÏßÄ ÌÅ¨Í∏∞
        max_zernike_order: ÏµúÎåÄ Zernike Ï∞®Ïàò
    """
    
    print("="*80)
    print("üî¨ ÏàúÏàò ÏòµÌã∞ÎßàÏù¥Ï†Ä Í∏∞Î∞ò ÏàòÏ∞® Î≥¥Ï†ï ÌÖåÏä§Ìä∏")
    print("="*80)
    print(f"ÏÑ§Ï†ï:")
    print(f"  ÏòµÌã∞ÎßàÏù¥Ï†Ä: {optimizer_name}")
    print(f"  ÌïôÏäµÎ•†: {learning_rate}")
    print(f"  ÏµúÎåÄ Î∞òÎ≥µ: {max_iterations}")
    print(f"  ÏàòÎ†¥ ÏûÑÍ≥ÑÍ∞í: {convergence_threshold}")
    print(f"  Ïù¥ÎØ∏ÏßÄ ÌÅ¨Í∏∞: {image_size}x{image_size}")
    print(f"  Zernike Ï∞®Ïàò: {max_zernike_order}")
    print("="*80)
    
    # ÏãúÎÆ¨Î†àÏù¥ÌÑ∞ Ï¥àÍ∏∞Ìôî
    if SIMULATOR_TYPE == 'point_source':
        simulator = PointSourceSimulator(
            grid_size=image_size,
            propagation_distance=1e-3,
            max_zernike_modes=max_zernike_order,
            target_position=None
        )
        print(f"‚úÖ Ï†êÍ¥ëÏõê ÏãúÎÆ¨Î†àÏù¥ÌÑ∞ Ï¥àÍ∏∞Ìôî ÏôÑÎ£å")
    else:
        simulator = BeamSimulator(
            image_size=(image_size, image_size),
            laser_to_slm_distance=0.05,
            slm_to_camera_distance=0.05,
            device='cuda' if torch.cuda.is_available() else 'cpu'
        )
        print(f"‚úÖ Îπî ÏãúÎÆ¨Î†àÏù¥ÌÑ∞ Ï¥àÍ∏∞Ìôî ÏôÑÎ£å")
    
    # Ï¥àÍ∏∞ ÏãúÎÆ¨Î†àÏù¥ÏÖò (ÎûúÎç§ ÏàòÏ∞® ÏÉùÏÑ±)
    if SIMULATOR_TYPE == 'point_source':
        initial_field = simulator.reset()
        initial_aberration = simulator.initial_aberration.clone()
        
        # Ï¥àÍ∏∞ Strehl ratio Í≥ÑÏÇ∞
        initial_metrics = simulator.calculate_strehl_ratio(initial_field)
        initial_strehl = initial_metrics['strehl_ratio']
    else:
        simulator._initialize_system_aberration()
        initial_result = simulator.simulate_complete_optical_path()
        initial_aberration = simulator.system_aberration_coeffs.clone().squeeze()
        
        # Ïù¥ÏÉÅÏ†ÅÏù∏ PSF
        ideal_psf = simulator.get_ideal_psf()
        initial_psf = initial_result['focus_psf']
        
        # Ï¥àÍ∏∞ Strehl ratio Í≥ÑÏÇ∞
        initial_metrics = simulator.calculate_strehl_ratio(initial_psf, ideal_psf)
        initial_strehl = initial_metrics['strehl_ratio']
    
    print(f"\nüìä Ï¥àÍ∏∞ ÏÉÅÌÉú:")
    print(f"  Ï¥àÍ∏∞ Strehl ratio: {initial_strehl:.6f}")
    print(f"  ÏãúÏä§ÌÖú ÏàòÏ∞® ÌÅ¨Í∏∞: {torch.norm(initial_aberration).item():.6f}")
    
    # ÏµúÏ†ÅÌôîÌï† Zernike Í≥ÑÏàò Ï¥àÍ∏∞Ìôî (numpy arrayÎ°ú ÏãúÏûë)
    initial_coeffs = np.zeros(max_zernike_order)
    
    print(f"\n‚úÖ {optimizer_name} ÏòµÌã∞ÎßàÏù¥Ï†Ä ÏÇ¨Ïö©")
    
    # ÏµúÏ†ÅÌôî Í∏∞Î°ù
    history = {
        'iterations': [],
        'strehl_ratios': [],
        'losses': [],
        'correction_norms': []
    }
    
    # Î∞òÎ≥µ Ïπ¥Ïö¥ÌÑ∞
    iteration_counter = [0]
    best_strehl = [initial_strehl]
    best_coeffs = [initial_coeffs.copy()]
    
    # Î™©Ï†Å Ìï®Ïàò Ï†ïÏùò (scipy.optimize.minimizeÏö©)
    def objective_function(coeffs):
        """
        Î™©Ï†Å Ìï®Ïàò: -Strehl ratio (ÏµúÏÜåÌôî)
        """
        iteration_counter[0] += 1
        
        # numpy arrayÎ•º torch tensorÎ°ú Î≥ÄÌôò
        coeffs_tensor = torch.tensor(coeffs, dtype=torch.float32)
        
        # ÏãúÎÆ¨Î†àÏù¥ÏÖò Ïã§Ìñâ
        if SIMULATOR_TYPE == 'point_source':
            # Ï¥àÍ∏∞ ÏàòÏ∞® + Î≥¥Ï†ï Í≥ÑÏàò (Í∞ïÌôîÌïôÏäµ ÌôòÍ≤ΩÍ≥º ÎèôÏùº)
            simulator.zernike_coefficients = simulator.initial_aberration + coeffs_tensor
            field = simulator._simulate()
            metrics = simulator.calculate_strehl_ratio(field)
            current_strehl = metrics['strehl_ratio']
        else:
            import torchOptics.zernike as tz
            zernike_array = tz.get_ZernikeArray(
                max_order=max_zernike_order,
                shape=(image_size, image_size),
                device='cpu'
            )
            
            phase_pattern = torch.zeros(image_size, image_size)
            for i in range(max_zernike_order):
                phase_pattern += coeffs_tensor[i] * zernike_array[i]
            
            result = simulator.simulate_complete_optical_path(phase_pattern.squeeze())
            current_psf = result['focus_psf']
            metrics = simulator.calculate_strehl_ratio(current_psf, ideal_psf)
            current_strehl = metrics['strehl_ratio']
        
        # Í∏∞Î°ù Ï†ÄÏû•
        correction_norm = np.linalg.norm(coeffs)
        history['iterations'].append(iteration_counter[0])
        history['strehl_ratios'].append(current_strehl)
        history['losses'].append(-current_strehl)
        history['correction_norms'].append(correction_norm)
        
        # ÏµúÍ≥† ÏÑ±Îä• ÏóÖÎç∞Ïù¥Ìä∏
        if current_strehl > best_strehl[0]:
            best_strehl[0] = current_strehl
            best_coeffs[0] = coeffs.copy()
        
        # ÏßÑÌñâ ÏÉÅÌô© Ï∂úÎ†• (10Î≤àÎßàÎã§)
        if iteration_counter[0] % 10 == 0 or current_strehl >= convergence_threshold:
            print(f"Iter {iteration_counter[0]:3d} | Strehl: {current_strehl:.6f} | "
                  f"Loss: {-current_strehl:.6f} | Correction norm: {correction_norm:.6f}")
        
        # Î™©Ï†Å Ìï®Ïàò Í∞í Î∞òÌôò (ÏµúÏÜåÌôîÌïòÎØÄÎ°ú ÏùåÏàò)
        return -current_strehl
    
    print(f"\nüöÄ ÏµúÏ†ÅÌôî ÏãúÏûë...\n")
    
    # scipy.optimize.minimize Ïã§Ìñâ
    result = minimize(
        objective_function,
        initial_coeffs,
        method=optimizer_name,
        options={
            'maxiter': max_iterations,
            'disp': False
        }
    )
    
    # ÏµúÏ¢Ö Í≤∞Í≥º
    final_coeffs = result.x
    final_strehl = -result.fun  # ÏµúÏÜåÌôîÌñàÏúºÎØÄÎ°ú Îã§Ïãú ÏùåÏàò Ï∑®Ìï®
    iteration = iteration_counter[0]
    
    print(f"\n" + "="*80)
    print(f"üìä ÏµúÏ†ÅÌôî Í≤∞Í≥º ÏöîÏïΩ")
    print(f"="*80)
    print(f"Ï¥àÍ∏∞ Strehl ratio: {initial_strehl:.6f}")
    print(f"ÏµúÏ¢Ö Strehl ratio: {final_strehl:.6f}")
    print(f"ÏµúÍ≥† Strehl ratio: {best_strehl[0]:.6f}")
    print(f"Í∞úÏÑ†ÎèÑ: {final_strehl - initial_strehl:.6f} ({(final_strehl - initial_strehl)/initial_strehl*100:+.2f}%)")
    print(f"Î∞òÎ≥µ ÌöüÏàò: {iteration}/{max_iterations}")
    print(f"ÏµúÏ¢Ö Î≥¥Ï†ï Í≥ÑÏàò ÌÅ¨Í∏∞: {np.linalg.norm(final_coeffs):.6f}")
    print(f"ÏàòÎ†¥ Ïó¨Î∂Ä: {result.success}")
    print(f"="*80)
    
    # Í≤∞Í≥º ÏãúÍ∞ÅÌôî
    visualize_optimization_results(history, initial_strehl, final_strehl, best_strehl[0])
    
    # Í≥ÑÏàò ÎπÑÍµê Î∂ÑÏÑù
    if SIMULATOR_TYPE != 'point_source':
        analyze_coefficient_comparison(initial_aberration, torch.tensor(best_coeffs[0]), max_zernike_order)
    
    return {
        'initial_strehl': initial_strehl,
        'final_strehl': final_strehl,
        'best_strehl': best_strehl[0],
        'iterations': iteration,
        'history': history,
        'best_coeffs': best_coeffs[0],
        'initial_aberration': initial_aberration,
        'success': result.success
    }


def visualize_optimization_results(history, initial_strehl, final_strehl, best_strehl):
    """ÏµúÏ†ÅÌôî Í≤∞Í≥º ÏãúÍ∞ÅÌôî"""
    fig, axes = plt.subplots(2, 2, figsize=(14, 10))
    
    # Strehl ratio ÏßÑÌñâ
    ax = axes[0, 0]
    ax.plot(history['iterations'], history['strehl_ratios'], 'b-', linewidth=2, label='Strehl ratio')
    ax.axhline(y=initial_strehl, color='r', linestyle='--', label=f'Initial: {initial_strehl:.4f}')
    ax.axhline(y=best_strehl, color='g', linestyle='--', label=f'Best: {best_strehl:.4f}')
    ax.set_xlabel('Iteration', fontsize=12)
    ax.set_ylabel('Strehl Ratio', fontsize=12)
    ax.set_title('Strehl Ratio ÏßÑÌñâ', fontsize=14, fontweight='bold')
    ax.legend()
    ax.grid(True, alpha=0.3)
    
    # ÏÜêÏã§ Ìï®Ïàò ÏßÑÌñâ
    ax = axes[0, 1]
    ax.plot(history['iterations'], history['losses'], 'r-', linewidth=2)
    ax.set_xlabel('Iteration', fontsize=12)
    ax.set_ylabel('Loss (-Strehl)', fontsize=12)
    ax.set_title('ÏÜêÏã§ Ìï®Ïàò ÏßÑÌñâ', fontsize=14, fontweight='bold')
    ax.grid(True, alpha=0.3)
    
    # Î≥¥Ï†ï Í≥ÑÏàò ÌÅ¨Í∏∞
    ax = axes[1, 0]
    ax.plot(history['iterations'], history['correction_norms'], 'g-', linewidth=2)
    ax.set_xlabel('Iteration', fontsize=12)
    ax.set_ylabel('Correction Norm', fontsize=12)
    ax.set_title('Î≥¥Ï†ï Í≥ÑÏàò ÌÅ¨Í∏∞ ÏßÑÌñâ', fontsize=14, fontweight='bold')
    ax.grid(True, alpha=0.3)
    
    # Î™©Ï†Å Ìï®Ïàò Í∞í (ÏÜêÏã§)
    ax = axes[1, 1]
    ax.plot(history['iterations'], np.abs(history['losses']), 'm-', linewidth=2)
    ax.set_xlabel('Iteration', fontsize=12)
    ax.set_ylabel('Objective Function Value', fontsize=12)
    ax.set_title('Î™©Ï†Å Ìï®Ïàò ÏßÑÌñâ (Ï†àÎåìÍ∞í)', fontsize=14, fontweight='bold')
    ax.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.savefig('optimizer_results.png', dpi=150, bbox_inches='tight')
    print(f"\n‚úÖ ÏµúÏ†ÅÌôî Í≤∞Í≥º Í∑∏ÎûòÌîÑ Ï†ÄÏû•: optimizer_results.png")
    plt.close()


def analyze_coefficient_comparison(initial_aberration, best_coeffs, max_zernike_order):
    """Ï¥àÍ∏∞ ÏàòÏ∞®ÏôÄ ÏµúÏ†Å Î≥¥Ï†ï Í≥ÑÏàò ÎπÑÍµê Î∂ÑÏÑù"""
    print(f"\n" + "="*80)
    print(f"üîç Í≥ÑÏàò ÎπÑÍµê Î∂ÑÏÑù")
    print(f"="*80)
    
    initial_aberr = initial_aberration.cpu().numpy() if hasattr(initial_aberration, 'cpu') else initial_aberration
    correction = best_coeffs.cpu().numpy() if hasattr(best_coeffs, 'cpu') else best_coeffs
    
    print(f"{'Index':<5} {'System':<12} {'Correction':<12} {'Direction':<12} {'Ratio':<12}")
    print(f"{'-'*60}")
    
    for i in range(1, max_zernike_order):  # Z0 Ï†úÏô∏
        if i < len(initial_aberr) and i < len(correction):
            system_val = initial_aberr[i]
            corr_val = correction[i]
            
            if abs(system_val) < 1e-6:
                direction = "N/A"
                ratio = "N/A"
            else:
                if system_val * corr_val < 0:
                    direction = "‚úÖ Cancel"
                    ratio = f"{abs(corr_val / system_val):.2f}x"
                else:
                    direction = "‚ùå Amplify"
                    ratio = f"{abs(corr_val / system_val):.2f}x"
            
            print(f"Z{i:<4} {system_val:<12.6f} {corr_val:<12.6f} {direction:<12} {ratio:<12}")
    
    print(f"="*80)


def run_multiple_tests():
    """Ïó¨Îü¨ ÏòµÌã∞ÎßàÏù¥Ï†ÄÎ°ú ÌÖåÏä§Ìä∏ Ïã§Ìñâ"""
    print("\n" + "="*80)
    print("üß™ Ïó¨Îü¨ ÏòµÌã∞ÎßàÏù¥Ï†Ä ÎπÑÍµê ÌÖåÏä§Ìä∏")
    print("="*80)
    
    optimizers_configs = [
        {'name': 'Nelder-Mead'},
        {'name': 'Powell'},
        {'name': 'COBYLA'},
    ]
    
    results = []
    
    for config in optimizers_configs:
        print(f"\n{'='*80}")
        print(f"ÌÖåÏä§Ìä∏: {config['name']}")
        print(f"{'='*80}")
        
        result = test_optimizer_correction(
            optimizer_name=config['name'],
            max_iterations=200
        )
        
        results.append({
            'optimizer': config['name'],
            'initial_strehl': result['initial_strehl'],
            'final_strehl': result['final_strehl'],
            'best_strehl': result['best_strehl'],
            'iterations': result['iterations'],
            'success': result['success']
        })
    
    # Í≤∞Í≥º ÎπÑÍµê Ï∂úÎ†•
    print(f"\n" + "="*80)
    print(f"üìä ÏòµÌã∞ÎßàÏù¥Ï†Ä ÎπÑÍµê Í≤∞Í≥º")
    print(f"="*80)
    print(f"{'Optimizer':<15} {'Initial':<10} {'Final':<10} {'Best':<10} {'Improvement':<12} {'Success':<10} {'Iters':<8}")
    print(f"{'-'*90}")
    
    for r in results:
        improvement = r['best_strehl'] - r['initial_strehl']
        success_str = "‚úÖ" if r['success'] else "‚ùå"
        print(f"{r['optimizer']:<15} {r['initial_strehl']:<10.6f} "
              f"{r['final_strehl']:<10.6f} {r['best_strehl']:<10.6f} "
              f"{improvement:<12.6f} {success_str:<10} {r['iterations']:<8d}")
    
    print(f"="*80)


if __name__ == "__main__":
    # Îã®Ïùº ÌÖåÏä§Ìä∏
    print("üî¨ Îã®Ïùº ÏòµÌã∞ÎßàÏù¥Ï†Ä ÌÖåÏä§Ìä∏ (Nelder-Mead)\n")
    
    test_optimizer_correction(
        optimizer_name='Nelder-Mead',
        max_iterations=200
    )
    
    # Ïó¨Îü¨ ÏòµÌã∞ÎßàÏù¥Ï†Ä ÎπÑÍµê ÌÖåÏä§Ìä∏
    print("\n\n")
    run_multiple_tests()

```

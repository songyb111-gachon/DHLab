```python
from gymnasium.wrappers import RecordVideo
import gymnasium as gym
import os
from stable_baselines3 import A2C

# 비디오 저장 경로 설정
video_folder = "./videos/"
os.makedirs(video_folder, exist_ok=True)

# 환경 생성 및 비디오 기록 래퍼 추가
env = gym.make("CartPole-v1", render_mode="rgb_array")  # render_mode 설정
env = RecordVideo(env, video_folder=video_folder, episode_trigger=lambda episode_id: True)

# 모델 생성 및 학습 (GPU 활성화: device="cuda")
model = A2C("MlpPolicy", env, verbose=1, device="cuda")  # CUDA 설정
model.learn(total_timesteps=1000000)

# 학습된 모델 실행
obs, info = env.reset()  # Gymnasium 반환값 처리
for _ in range(1000):
    action, _state = model.predict(obs, deterministic=True)
    obs, reward, done, info = env.step(action)
    if done:
        obs, info = env.reset()

env.close()
print(f"비디오가 {video_folder}에 저장되었습니다.")
```

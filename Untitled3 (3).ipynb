```python
# 빔 시뮬레이터 테스트 스크립트
# 완전한 광학 경로: 레이저 → SLM → 카메라 시뮬레이션 테스트

import torch
import numpy as np
import matplotlib.pyplot as plt
from beam_simulator import BeamSimulator

def test_complete_optical_path():
    # 완전한 광학 경로 시뮬레이션 테스트
    print("=== 완전한 광학 경로 시뮬레이션 테스트 ===")
    
    # 디바이스 설정
    device = 'cuda' if torch.cuda.is_available() else 'cpu'
    print(f"사용 디바이스: {device}")
    
    # 빔 시뮬레이터 초기화
    simulator = BeamSimulator(
        image_size=(512, 512),  # 512x512 기준
        pixel_size=(6.3e-6, 6.3e-6),
        wavelength=638e-9,
        laser_to_slm_distance=0.05,  # 5cm
        slm_to_camera_distance=0.05,  # 5cm
        device=device
    )
    
    print("빔 시뮬레이터 초기화 완료")
    
    # 시스템 수차 정보 출력
    aberration_info = simulator.get_system_aberration_info()
    print(f"시스템 수차 총 크기: {aberration_info['total_aberration_magnitude']:.4f}")
    
    # 1. SLM 없이 통과하는 경우 (기본 테스트)
    print("\n1. SLM 통과 테스트 (패턴 없음)")
    result_no_pattern = simulator.simulate_complete_optical_path()
    
    print("카메라 메트릭:")
    for key, value in result_no_pattern['metrics'].items():
        print(f"  {key}: {value:.4f}")
    
    # 2. SLM에 랜덤 위상 패턴 적용
    print("\n2. SLM 랜덤 위상 패턴 테스트")
    slm_pattern = torch.randn(512, 512, device=device) * 0.5  # 랜덤 위상 패턴
    result_with_pattern = simulator.simulate_complete_optical_path(slm_pattern)
    
    print("카메라 메트릭 (패턴 적용 후):")
    for key, value in result_with_pattern['metrics'].items():
        print(f"  {key}: {value:.4f}")
    
    # 3. 시각화
    print("\n3. 광학 경로 시각화")
    simulator.visualize_complete_path(result_no_pattern, "No SLM Pattern")
    simulator.visualize_complete_path(result_with_pattern, "With SLM Pattern")
    
    return result_no_pattern, result_with_pattern

def test_perfect_optical_system():
    # 수차가 없는 완벽한 광학 시스템 테스트
    print("\n=== 완벽한 광학 시스템 테스트 (수차 없음) ===")
    
    device = 'cuda' if torch.cuda.is_available() else 'cpu'
    
    # 완벽한 시스템 시뮬레이터 생성
    simulator_perfect = BeamSimulator(
        image_size=(512, 512),
        laser_to_slm_distance=0.05,
        slm_to_camera_distance=0.05,
        device=device
    )
    
    # 시스템 수차를 완전히 제거 (0으로 설정)
    simulator_perfect.system_aberration_coeffs = torch.zeros_like(
        simulator_perfect.system_aberration_coeffs
    )
    
    print("시스템 수차를 완전히 제거했습니다.")
    
    # 완벽한 시스템에서의 시뮬레이션
    result_perfect = simulator_perfect.simulate_complete_optical_path()
    
    print("완벽한 시스템 메트릭:")
    for key, value in result_perfect['metrics'].items():
        print(f"  {key}: {value:.6f}")
    
    # SLM 패턴 없이 완벽한 시스템
    print("\n완벽한 시스템 + SLM 패턴 없음:")
    result_perfect_no_slm = simulator_perfect.simulate_complete_optical_path()
    
    print("카메라 메트릭:")
    for key, value in result_perfect_no_slm['metrics'].items():
        print(f"  {key}: {value:.6f}")
    
    # SLM에 랜덤 패턴 적용 (완벽한 시스템에서)
    print("\n완벽한 시스템 + SLM 랜덤 패턴:")
    slm_pattern = torch.randn(512, 512, device=device) * 0.3  # 작은 위상 변조
    result_perfect_with_slm = simulator_perfect.simulate_complete_optical_path(slm_pattern)
    
    print("카메라 메트릭 (SLM 패턴 적용 후):")
    for key, value in result_perfect_with_slm['metrics'].items():
        print(f"  {key}: {value:.6f}")
    
    # 시각화
    print("\n완벽한 시스템 시각화:")
    simulator_perfect.visualize_complete_path(result_perfect_no_slm, "Perfect System - No SLM")
    simulator_perfect.visualize_complete_path(result_perfect_with_slm, "Perfect System - With SLM")
    
    return result_perfect_no_slm, result_perfect_with_slm

def test_comparison_perfect_vs_aberrated():
    # 완벽한 시스템 vs 수차가 있는 시스템 비교
    print("\n=== 완벽한 시스템 vs 수차가 있는 시스템 비교 ===")
    
    device = 'cuda' if torch.cuda.is_available() else 'cpu'
    
    # 완벽한 시스템
    simulator_perfect = BeamSimulator(
        image_size=(512, 512),
        device=device
    )
    simulator_perfect.system_aberration_coeffs = torch.zeros_like(
        simulator_perfect.system_aberration_coeffs
    )
    
    # 수차가 있는 시스템
    simulator_aberrated = BeamSimulator(
        image_size=(512, 512),
        device=device
    )
    
    # 각각 시뮬레이션
    result_perfect = simulator_perfect.simulate_complete_optical_path()
    result_aberrated = simulator_aberrated.simulate_complete_optical_path()
    
    # 비교 출력
    print("메트릭 비교:")
    print(f"{'메트릭':<20} {'완벽한 시스템':<15} {'수차가 있는 시스템':<20} {'차이':<10}")
    print("-" * 70)
    
    for key in result_perfect['metrics'].keys():
        perfect_val = result_perfect['metrics'][key]
        aberrated_val = result_aberrated['metrics'][key]
        difference = perfect_val - aberrated_val
        
        print(f"{key:<20} {perfect_val:<15.6f} {aberrated_val:<20.6f} {difference:<10.6f}")
    
    # 성능 저하율 계산
    perfect_beam_quality = result_perfect['metrics']['beam_quality']
    aberrated_beam_quality = result_aberrated['metrics']['beam_quality']
    degradation = (1 - aberrated_beam_quality / perfect_beam_quality) * 100
    
    print(f"\n수차로 인한 Beam Quality 성능 저하: {degradation:.2f}%")
    
    # 시각화 비교
    print("\n시각화 비교:")
    simulator_perfect.visualize_complete_path(result_perfect, "Perfect System")
    simulator_aberrated.visualize_complete_path(result_aberrated, "Aberrated System")
    
    return result_perfect, result_aberrated

def test_system_aberration_impact():
    # 시스템 수차의 영향 테스트
    print("\n=== 시스템 수차 영향 테스트 ===")
    
    device = 'cuda' if torch.cuda.is_available() else 'cpu'
    
    # 수차가 없는 경우 (이상적인 경우)
    simulator_ideal = BeamSimulator(
        image_size=(512, 512),
        device=device
    )
    
    # 시스템 수차를 0으로 설정
    simulator_ideal.system_aberration_coeffs = torch.zeros_like(
        simulator_ideal.system_aberration_coeffs
    )
    
    result_ideal = simulator_ideal.simulate_complete_optical_path()
    
    # 수차가 있는 경우 (현재 설정)
    simulator_aberrated = BeamSimulator(
        image_size=(512, 512),
        device=device
    )
    
    result_aberrated = simulator_aberrated.simulate_complete_optical_path()
    
    # 비교
    print("이상적인 경우 (수차 없음):")
    for key, value in result_ideal['metrics'].items():
        print(f"  {key}: {value:.4f}")
    
    print("\n수차가 있는 경우:")
    for key, value in result_aberrated['metrics'].items():
        print(f"  {key}: {value:.4f}")
    
    # Beam Quality 비교
    ideal_beam_quality = result_ideal['metrics']['beam_quality']
    aberrated_beam_quality = result_aberrated['metrics']['beam_quality']
    
    print(f"\nBeam Quality 비교:")
    print(f"  이상적인 경우: {ideal_beam_quality:.4f}")
    print(f"  수차가 있는 경우: {aberrated_beam_quality:.4f}")
    print(f"  수차로 인한 성능 저하: {(1 - aberrated_beam_quality/ideal_beam_quality)*100:.2f}%")

def test_different_resolutions():
    # 다양한 해상도에서의 시뮬레이션 테스트
    print("\n=== 해상도별 시뮬레이션 테스트 ===")
    
    device = 'cuda' if torch.cuda.is_available() else 'cpu'
    
    resolutions = [(128, 128), (256, 256), (512, 512)]
    
    for resolution in resolutions:
        print(f"\n해상도: {resolution[0]}x{resolution[1]}")
        
        simulator = BeamSimulator(
            image_size=resolution,
            device=device
        )
        
        result = simulator.simulate_complete_optical_path()
        
        print(f"  Beam Quality: {result['metrics']['beam_quality']:.4f}")
        print(f"  Peak intensity: {result['metrics']['peak_intensity']:.4f}")
        print(f"  Sharpness: {result['metrics']['sharpness']:.4f}")
        
        # 시스템 수차 정보
        aberration_info = simulator.get_system_aberration_info()
        print(f"  시스템 수차 총 크기: {aberration_info['total_aberration_magnitude']:.4f}")

def test_optimal_aberration_correction():
    # 최적의 수차 보정 테스트
    print("\n=== 최적의 수차 보정 테스트 ===")
    
    device = 'cuda' if torch.cuda.is_available() else 'cpu'
    
    # 수차가 있는 시스템 생성
    simulator = BeamSimulator(
        image_size=(512, 512),
        laser_to_slm_distance=0.05,
        slm_to_camera_distance=0.05,
        device=device
    )
    
    # 시스템 수차 정보 출력
    aberration_info = simulator.get_system_aberration_info()
    print(f"시스템 수차 총 크기: {aberration_info['total_aberration_magnitude']:.4f}")
    print(f"시스템 수차 계수 (처음 5개): {aberration_info['coefficients'][:5]}")
    
    # 1. 수차 보정 없이 (원래 상태)
    print("\n1. 수차 보정 없이 (원래 상태)")
    result_no_correction = simulator.simulate_complete_optical_path()
    
    print("카메라 메트릭 (보정 없음):")
    for key, value in result_no_correction['metrics'].items():
        print(f"  {key}: {value:.6f}")
    
    # 2. 최적의 수차 보정 적용 (시스템 수차의 음수)
    print("\n2. 최적의 수차 보정 적용")
    
    # 최적 보정: 시스템 수차의 음수
    optimal_correction_coeffs = -simulator.system_aberration_coeffs.squeeze()
    print(f"최적 보정 계수 (처음 5개): {optimal_correction_coeffs[:5].cpu().numpy()}")
    
    # Zernike 계수를 SLM 패턴으로 변환
    import sys
    sys.path.append('./torchOptics')
    import torchOptics.zernike as tz
    
    # Zernike 패턴 생성
    zernike_array = tz.get_ZernikeArray(
        max_order=15, 
        shape=simulator.image_size, 
        device=device
    )
    
    # 최적 보정 패턴 계산
    optimal_slm_pattern = torch.zeros(simulator.image_size, device=device)
    for i in range(len(optimal_correction_coeffs)):
        optimal_slm_pattern += optimal_correction_coeffs[i] * zernike_array[i]
    
    # 최적 보정 적용
    result_optimal_correction = simulator.simulate_complete_optical_path(optimal_slm_pattern)
    
    print("카메라 메트릭 (최적 보정 적용):")
    for key, value in result_optimal_correction['metrics'].items():
        print(f"  {key}: {value:.6f}")
    
    # 🔍 Beam Quality 구성 요소 분석
    print("\n🔍 Beam Quality 구성 요소 분석:")
    
    def analyze_beam_quality_components(metrics, label):
        print(f"\n{label}:")
        peak_int = metrics['peak_intensity']
        total_energy = metrics['total_energy']
        center_conc = metrics['center_concentration']
        sharpness = metrics['sharpness']
        symmetry_err = metrics['symmetry_error']
        
        # 계산 과정 재현
        relative_peak = min(peak_int / total_energy * 10000, 1.0)
        smoothness_score = min(1.0 / (1.0 + sharpness * 100), 1.0)
        symmetry_score = 1.0 / (1.0 + symmetry_err)
        
        print(f"  relative_peak_intensity: {relative_peak:.6f} (weight: 0.3)")
        print(f"  symmetry_score: {symmetry_score:.6f} (weight: 0.3)")
        print(f"  center_concentration: {center_conc:.6f} (weight: 0.2)")
        print(f"  smoothness_score: {smoothness_score:.6f} (weight: 0.2) [낮은 sharpness = 좋음]")
        
        calculated_quality = (
            relative_peak * 0.3 +
            symmetry_score * 0.3 +
            center_conc * 0.2 +
            smoothness_score * 0.2
        )
        print(f"  계산된 Beam Quality: {calculated_quality:.6f}")
        print(f"  실제 Beam Quality: {metrics['beam_quality']:.6f}")
    
    analyze_beam_quality_components(result_no_correction['metrics'], "보정 없음")
    analyze_beam_quality_components(result_optimal_correction['metrics'], "최적 보정")
    
    # 🔍 완벽한 시스템과 비교
    print("\n🔍 완벽한 시스템(수차 없음)과 비교:")
    
    # 완벽한 시스템 생성 (수차 제거)
    perfect_simulator = BeamSimulator(
        image_size=(512, 512),
        laser_to_slm_distance=0.05,
        slm_to_camera_distance=0.05,
        device=device
    )
    perfect_simulator.system_aberration_coeffs = torch.zeros_like(
        perfect_simulator.system_aberration_coeffs
    )
    
    result_perfect = perfect_simulator.simulate_complete_optical_path()
    
    print("완벽한 시스템 메트릭:")
    for key, value in result_perfect['metrics'].items():
        print(f"  {key}: {value:.6f}")
    
    analyze_beam_quality_components(result_perfect['metrics'], "완벽한 시스템")
    
    # 완벽한 시스템 vs 최적 보정 비교
    perfect_quality = result_perfect['metrics']['beam_quality']
    corrected_quality = result_optimal_correction['metrics']['beam_quality']
    correction_effectiveness = corrected_quality / perfect_quality * 100
    
    print(f"\n🎯 보정 효과 분석:")
    print(f"  완벽한 시스템 Beam Quality: {perfect_quality:.6f}")
    print(f"  최적 보정 후 Beam Quality: {corrected_quality:.6f}")
    print(f"  보정 효과: {correction_effectiveness:.2f}% (완벽 대비)")
    
    if correction_effectiveness >= 95:
        print("  ✅ 거의 완벽한 보정!")
    elif correction_effectiveness >= 85:
        print("  ✅ 매우 좋은 보정!")
    elif correction_effectiveness >= 70:
        print("  ⚠️ 괜찮은 보정")
    else:
        print("  ❌ 보정 부족")
    
    # 3. 성능 비교
    print("\n3. 성능 비교")
    print(f"{'메트릭':<25} {'보정 없음':<15} {'최적 보정':<15} {'개선도':<15}")
    print("-" * 75)
    
    for key in result_no_correction['metrics'].keys():
        no_corr_val = result_no_correction['metrics'][key]
        opt_corr_val = result_optimal_correction['metrics'][key]
        improvement = opt_corr_val - no_corr_val
        improvement_pct = (improvement / no_corr_val) * 100 if no_corr_val != 0 else 0
        
        print(f"{key:<25} {no_corr_val:<15.6f} {opt_corr_val:<15.6f} {improvement_pct:<15.2f}%")
    
    # 4. Beam Quality 개선 요약
    original_quality = result_no_correction['metrics']['beam_quality']
    corrected_quality = result_optimal_correction['metrics']['beam_quality']
    quality_improvement = corrected_quality - original_quality
    quality_improvement_pct = (quality_improvement / original_quality) * 100
    
    print(f"\n🎯 Beam Quality 개선 요약:")
    print(f"  원래 Beam Quality: {original_quality:.6f}")
    print(f"  보정 후 Beam Quality: {corrected_quality:.6f}")
    print(f"  개선도: {quality_improvement:.6f} (+{quality_improvement_pct:.2f}%)")
    
    # 5. 시각화
    print("\n4. 시각화 비교")
    simulator.visualize_complete_path(result_no_correction, "No Correction")
    simulator.visualize_complete_path(result_optimal_correction, "Optimal Correction")
    
    return result_no_correction, result_optimal_correction

def test_different_setups_for_success_criteria():
    # 다양한 셋업에서 성공 조건 적절성 테스트
    print("\n=== 다양한 셋업에서 성공 조건 적절성 테스트 ===")
    
    device = 'cuda' if torch.cuda.is_available() else 'cpu'
    
    # 현재 목표값들
    TARGET_BEAM_QUALITY = 0.5
    TARGET_IMPROVEMENT = 0.05
    
    print(f"현재 목표 설정:")
    print(f"  TARGET_BEAM_QUALITY: {TARGET_BEAM_QUALITY}")
    print(f"  TARGET_IMPROVEMENT: {TARGET_IMPROVEMENT}")
    
    # 다양한 셋업 테스트
    setups = [
        {"name": "기본 셋업 (5cm+5cm)", "laser_to_slm": 0.05, "slm_to_camera": 0.05},
        {"name": "짧은 거리 (2cm+2cm)", "laser_to_slm": 0.02, "slm_to_camera": 0.02},
        {"name": "긴 거리 (10cm+10cm)", "laser_to_slm": 0.10, "slm_to_camera": 0.10},
        {"name": "비대칭 (3cm+7cm)", "laser_to_slm": 0.03, "slm_to_camera": 0.07},
    ]
    
    results_summary = []
    
    for setup in setups:
        print(f"\n--- {setup['name']} ---")
        
        # 여러 번 테스트 (랜덤 수차)
        perfect_qualities = []
        aberrated_qualities = []
        corrected_qualities = []
        improvements = []
        success_rates = []
        
        num_tests = 5  # 각 셋업마다 5번 테스트
        
        for test_i in range(num_tests):
            # 수차가 있는 시스템
            simulator = BeamSimulator(
                image_size=(512, 512),
                laser_to_slm_distance=setup["laser_to_slm"],
                slm_to_camera_distance=setup["slm_to_camera"],
                device=device
            )
            
            # 완벽한 시스템
            perfect_simulator = BeamSimulator(
                image_size=(512, 512),
                laser_to_slm_distance=setup["laser_to_slm"],
                slm_to_camera_distance=setup["slm_to_camera"],
                device=device
            )
            perfect_simulator.system_aberration_coeffs = torch.zeros_like(
                perfect_simulator.system_aberration_coeffs
            )
            
            # 결과 계산
            result_perfect = perfect_simulator.simulate_complete_optical_path()
            result_aberrated = simulator.simulate_complete_optical_path()
            
            # 최적 보정
            optimal_correction_coeffs = -simulator.system_aberration_coeffs.squeeze()
            
            import sys
            sys.path.append('./torchOptics')
            import torchOptics.zernike as tz
            
            zernike_array = tz.get_ZernikeArray(
                max_order=15, 
                shape=simulator.image_size, 
                device=device
            )
            
            optimal_slm_pattern = torch.zeros(simulator.image_size, device=device)
            for i in range(len(optimal_correction_coeffs)):
                optimal_slm_pattern += optimal_correction_coeffs[i] * zernike_array[i]
            
            result_corrected = simulator.simulate_complete_optical_path(optimal_slm_pattern)
            
            # 메트릭 수집
            perfect_quality = result_perfect['metrics']['beam_quality']
            aberrated_quality = result_aberrated['metrics']['beam_quality']
            corrected_quality = result_corrected['metrics']['beam_quality']
            improvement = corrected_quality - aberrated_quality
            
            perfect_qualities.append(perfect_quality)
            aberrated_qualities.append(aberrated_quality)
            corrected_qualities.append(corrected_quality)
            improvements.append(improvement)
            
            # 성공 조건 체크
            success = (corrected_quality >= TARGET_BEAM_QUALITY and improvement >= TARGET_IMPROVEMENT)
            success_rates.append(success)
        
        # 통계 계산
        avg_perfect = np.mean(perfect_qualities)
        avg_aberrated = np.mean(aberrated_qualities)
        avg_corrected = np.mean(corrected_qualities)
        avg_improvement = np.mean(improvements)
        success_rate = np.mean(success_rates) * 100
        
        print(f"  완벽한 시스템 평균 Beam Quality: {avg_perfect:.4f} ± {np.std(perfect_qualities):.4f}")
        print(f"  수차 있는 시스템 평균 Beam Quality: {avg_aberrated:.4f} ± {np.std(aberrated_qualities):.4f}")
        print(f"  최적 보정 후 평균 Beam Quality: {avg_corrected:.4f} ± {np.std(corrected_qualities):.4f}")
        print(f"  평균 개선도: {avg_improvement:.4f} ± {np.std(improvements):.4f}")
        print(f"  목표 달성률: {success_rate:.1f}% ({sum(success_rates)}/{num_tests})")
        
        # 적절성 평가
        if success_rate >= 80:
            status = "✅ 매우 적절"
        elif success_rate >= 60:
            status = "⚠️ 적절"
        elif success_rate >= 40:
            status = "⚠️ 다소 어려움"
        else:
            status = "❌ 너무 어려움"
        
        print(f"  평가: {status}")
        
        results_summary.append({
            'setup': setup['name'],
            'avg_perfect': avg_perfect,
            'avg_aberrated': avg_aberrated,
            'avg_corrected': avg_corrected,
            'avg_improvement': avg_improvement,
            'success_rate': success_rate,
            'status': status
        })
    
    # 전체 요약
    print(f"\n🎯 전체 셋업 요약:")
    print(f"{'셋업':<20} {'완벽한 BQ':<12} {'수차 BQ':<12} {'보정 BQ':<12} {'개선도':<10} {'성공률':<10} {'평가'}")
    print("-" * 90)
    
    for result in results_summary:
        print(f"{result['setup']:<20} {result['avg_perfect']:<12.4f} {result['avg_aberrated']:<12.4f} "
              f"{result['avg_corrected']:<12.4f} {result['avg_improvement']:<10.4f} "
              f"{result['success_rate']:<10.1f}% {result['status']}")
    
    # 권장사항
    overall_success_rate = np.mean([r['success_rate'] for r in results_summary])
    print(f"\n💡 권장사항:")
    print(f"  전체 평균 성공률: {overall_success_rate:.1f}%")
    
    if overall_success_rate >= 70:
        print("  ✅ 현재 목표 설정이 적절합니다!")
    elif overall_success_rate >= 50:
        print("  ⚠️ 현재 목표가 다소 어려울 수 있습니다.")
        print(f"  제안: TARGET_BEAM_QUALITY를 {TARGET_BEAM_QUALITY-0.05:.2f}로 낮추거나")
        print(f"       TARGET_IMPROVEMENT를 {TARGET_IMPROVEMENT-0.01:.2f}로 낮추세요.")
    else:
        print("  ❌ 현재 목표가 너무 어렵습니다!")
        print(f"  제안: TARGET_BEAM_QUALITY를 {TARGET_BEAM_QUALITY-0.1:.2f}로 낮추고")
        print(f"       TARGET_IMPROVEMENT를 {TARGET_IMPROVEMENT-0.02:.2f}로 낮추세요.")
    
    return results_summary

def test_different_distances():
    # 다양한 거리에서의 시뮬레이션 테스트
    print("\n=== 거리별 시뮬레이션 테스트 ===")
    
    device = 'cuda' if torch.cuda.is_available() else 'cpu'
    
    distances = [0.02, 0.05, 0.1, 0.2]  # 2cm, 5cm, 10cm, 20cm
    
    for distance in distances:
        print(f"\n거리: {distance*100:.0f}cm")
        
        simulator = BeamSimulator(
            image_size=(512, 512),
            laser_to_slm_distance=distance/2,
            slm_to_camera_distance=distance/2,
            device=device
        )
        
        result = simulator.simulate_complete_optical_path()
        
        print(f"  Beam Quality: {result['metrics']['beam_quality']:.4f}")
        print(f"  Peak intensity: {result['metrics']['peak_intensity']:.4f}")

# 메인 테스트 실행
try:
    # 완전한 광학 경로 테스트
    result_no_pattern, result_with_pattern = test_complete_optical_path()
    
    # 완벽한 시스템 테스트
    result_perfect_no_slm, result_perfect_with_slm = test_perfect_optical_system()
    
    # 시스템 수차 영향 테스트
    test_system_aberration_impact()
    
    # 완벽한 시스템 vs 수차가 있는 시스템 비교
    result_perfect, result_aberrated = test_comparison_perfect_vs_aberrated()
    
    # 해상도별 테스트
    test_different_resolutions()
    
    # 거리별 테스트
    test_different_distances()
    
    # 최적의 수차 보정 테스트
    result_no_correction, result_optimal_correction = test_optimal_aberration_correction()
    
    # 다양한 셋업에서 성공 조건 적절성 테스트
    setup_results = test_different_setups_for_success_criteria()
    
    print("\n=== 모든 테스트 완료 ===")
    
    # 최종 요약
    print("\n=== 최종 요약 ===")
    print(f"완벽한 시스템 Beam Quality: {result_perfect['metrics']['beam_quality']:.6f}")
    print(f"수차가 있는 시스템 Beam Quality: {result_aberrated['metrics']['beam_quality']:.6f}")
    print(f"최적 보정 후 Beam Quality: {result_optimal_correction['metrics']['beam_quality']:.6f}")
    print(f"성능 저하율 (완벽 vs 수차): {(1 - result_aberrated['metrics']['beam_quality'] / result_perfect['metrics']['beam_quality']) * 100:.2f}%")
    print(f"보정 효과 (수차 vs 최적보정): {((result_optimal_correction['metrics']['beam_quality'] - result_aberrated['metrics']['beam_quality']) / result_aberrated['metrics']['beam_quality']) * 100:.2f}%")
    
except Exception as e:
    print(f"테스트 중 오류 발생: {e}")
    import traceback
    traceback.print_exc()
```

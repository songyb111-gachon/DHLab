```python
# ============================================================
#     ( 4 )
# 1)    
# 2)  (threshold)   
# 3)     (   )
# 4) Smooth  ( edge smoothing) + MFS( ) 
#
# :
# - matplotlib  (/)    "" 
# ============================================================

import math
import numpy as np
import matplotlib.pyplot as plt
from dataclasses import dataclass
from scipy.interpolate import RectBivariateSpline
from scipy import ndimage as ndi


# ============================================================
#   (  )
# ============================================================

@dataclass
class MaskGenConfig:
    # ---------- A.  /  ----------
    unit_px: int = 256                 #  : unit_px x unit_px
    supercell_n_units: int = 4         # 1x4 (  4 )

    # ---------- B.    ----------
    ctrl_grid: int = 3                 #     (3 -> 3x3)
    spline_order: int = 2              # RectBivariateSpline (2=quadratic, 3=cubic)

    # ---------- C. (threshold) ----------
    threshold: float = 0.0             #  = (field > threshold)

    # ---------- D. (   ) ----------
    buffer_px: int = 6                 #   frame 0   

    # ---------- E.  smoothing (threshold ) ----------
    surface_gaussian_sigma: float = 1.0   #    Gaussian ( )
                                          # 0 

    # ---------- F. MFS(  )  ----------
    mfs_px: int = 10                   # morphology (   )
    min_island_area_px: int = 50       #   ()  ( )

    # ---------- G. edge smoothing ( Smooth ) ----------
    # SDF(signed distance field) Gaussian     
    mask_edge_smooth_sigma: float = 6.0    #    2~6   (0 )
    mask_edge_level: float = 0.0          # 0   
                                          # + (), - ()

    # edge smoothing    :
    # - "after_mfs": MFS    edge smoothing ()
    # - "before_mfs": edge smoothing  MFS (MFS smoothing    )
    edge_smooth_stage: str = "after_mfs"  # "after_mfs"  "before_mfs"

    # edge smoothing      
    final_remove_islands: bool = True

    # ---------- H.   ----------
    rng_seed: int = 42
    ctrl_value_range: tuple[float, float] = (-1.0, 1.0)  #    

    # ---------- I.  ----------
    show_intermediate_for_first: bool = True
    n_units_to_generate: int = 80
    show_all_units_grid: bool = True
    show_supercells: bool = True


# ============================================================
#  
# ============================================================

def _make_quadrant_surface(
    ctrl: np.ndarray,
    qn: int,
    *,
    spline_order: int,
    smooth_sigma: float,
) -> np.ndarray:
    """
     (Quadrant)    
    - ctrl: (g,g)   (: 3x3)
    - qn:   (= unit_px//2)
    """
    g = ctrl.shape[0]
    if ctrl.shape != (g, g):
        raise ValueError("ctrl must be a square 2D array, e.g. (3,3).")

    gx = np.linspace(0.0, 1.0, g)
    gy = np.linspace(0.0, 1.0, g)

    k = int(spline_order)
    k = max(1, min(k, g - 1))  # k <= g-1 

    spline = RectBivariateSpline(gx, gy, ctrl, kx=k, ky=k)

    xq = np.linspace(0.0, 1.0, qn)
    yq = np.linspace(0.0, 1.0, qn)
    S = spline(xq, yq)  # (qn, qn)

    if smooth_sigma and smooth_sigma > 0:
        S = ndi.gaussian_filter(S, sigma=float(smooth_sigma), mode="reflect")

    return S


def _enforce_fourfold_symmetry(quadrant: np.ndarray) -> np.ndarray:
    """
      4 (C4)   
    - quadrant: (qn,qn)
    - : (2qn,2qn) = (unit_px, unit_px)
    """
    q = quadrant
    left = np.fliplr(q)
    top = np.concatenate([left, q], axis=1)
    bottom = np.flipud(top)
    full = np.concatenate([bottom, top], axis=0)
    return full


def _apply_buffer(mask: np.ndarray, buffer_px: int) -> np.ndarray:
    """
       0     ( )
    """
    if buffer_px <= 0:
        return mask.astype(np.uint8)

    m = mask.astype(np.uint8).copy()
    b = int(buffer_px)
    m[:b, :] = 0
    m[-b:, :] = 0
    m[:, :b] = 0
    m[:, -b:] = 0
    return m


def _mfs_filter(mask: np.ndarray, mfs_px: int, min_island_area_px: int) -> np.ndarray:
    """
      (MFS)   :
    - opening:   
    - closing:  / 
    -  () 
    """
    m = mask.astype(bool)

    if mfs_px and mfs_px > 1:
        r = max(1, int(np.round(mfs_px / 2)))
        se = ndi.generate_binary_structure(2, 1)
        se = ndi.iterate_structure(se, r)

        m = ndi.binary_opening(m, structure=se)
        m = ndi.binary_closing(m, structure=se)

    m = m.astype(np.uint8)

    labeled, n = ndi.label(m)
    if n > 0 and min_island_area_px and min_island_area_px > 0:
        counts = np.bincount(labeled.ravel())
        remove = np.where(counts < int(min_island_area_px))[0]
        remove = remove[remove != 0]
        for lab in remove:
            m[labeled == lab] = 0

    return m.astype(np.uint8)


def _smooth_mask_edges(mask: np.ndarray, *, sigma: float, level: float) -> np.ndarray:
    """
        ( Smooth  ):
      mask -> signed distance field(SDF) -> Gaussian -> threshold(level)

    - sigma      (2~6 )
    - level=0   
      level>0:  ()
      level<0:  ()
    """
    if sigma is None or sigma <= 0:
        return mask.astype(np.uint8)

    m = mask.astype(bool)

    din = ndi.distance_transform_edt(m)
    dout = ndi.distance_transform_edt(~m)
    sdf = din - dout

    sdf_s = ndi.gaussian_filter(sdf, sigma=float(sigma), mode="reflect")
    return (sdf_s > float(level)).astype(np.uint8)


# ============================================================
#    API
# ============================================================

def generate_unit_cell_mask(
    ctrl_heights_quadrant: np.ndarray,
    cfg: MaskGenConfig,
    *,
    threshold: float | None = None,
    return_intermediates: bool = False,
):
    """
       ():
    1)    + 4 
    2) threshold 
    3) buffer 
    4) edge smoothing MFS / 
    5) MFS  + ()   
    """
    unit_px = int(cfg.unit_px)
    if unit_px % 2 != 0:
        raise ValueError("unit_px must be even to build 4-fold symmetry from a quadrant.")
    qn = unit_px // 2

    thr = cfg.threshold if threshold is None else float(threshold)

    quad = _make_quadrant_surface(
        ctrl_heights_quadrant,
        qn,
        spline_order=cfg.spline_order,
        smooth_sigma=cfg.surface_gaussian_sigma,
    )
    field = _enforce_fourfold_symmetry(quad)

    mask_thr = (field > thr).astype(np.uint8)
    mask_buffer = _apply_buffer(mask_thr, cfg.buffer_px)

    if cfg.edge_smooth_stage == "before_mfs":
        mask_smooth = _smooth_mask_edges(
            mask_buffer,
            sigma=cfg.mask_edge_smooth_sigma,
            level=cfg.mask_edge_level,
        )
        mask_mfs = _mfs_filter(mask_smooth, cfg.mfs_px, cfg.min_island_area_px)
        mask_final = mask_mfs
    else:
        mask_mfs = _mfs_filter(mask_buffer, cfg.mfs_px, cfg.min_island_area_px)
        mask_smooth = _smooth_mask_edges(
            mask_mfs,
            sigma=cfg.mask_edge_smooth_sigma,
            level=cfg.mask_edge_level,
        )
        mask_final = mask_smooth

        if cfg.final_remove_islands:
            # smoothing  morphology   ,   (  )
            mask_final = _mfs_filter(mask_final, mfs_px=1, min_island_area_px=cfg.min_island_area_px)

    if return_intermediates:
        return {
            "field": field,
            "mask_thr": mask_thr,
            "mask_buffer": mask_buffer,
            "mask_mfs": mask_mfs,
            "mask_smooth": mask_smooth,
            "mask_final": mask_final,
        }
    return mask_final


def build_supercell_1x4(unit_masks: list[np.ndarray], cfg: MaskGenConfig) -> np.ndarray:
    """
     4 (x)   1x4  
     : (unit_px, 4*unit_px)
    """
    if len(unit_masks) != int(cfg.supercell_n_units):
        raise ValueError(f"Need exactly {cfg.supercell_n_units} unit masks.")
    for m in unit_masks:
        if m.shape != (cfg.unit_px, cfg.unit_px):
            raise ValueError("All unit masks must be (unit_px, unit_px).")

    supercell = np.concatenate(unit_masks, axis=1)
    return supercell.astype(np.uint8)


# ============================================================
#   (  )
# ============================================================

def _show_intermediates(interm: dict, title_prefix: str = "Unit 0"):
    stages = [
        ("Spline field (2D)", interm["field"]),
        ("Threshold mask", interm["mask_thr"]),
        ("After buffer", interm["mask_buffer"]),
        ("After MFS", interm["mask_mfs"]),
        ("Edge-smoothed", interm["mask_smooth"]),
        ("Final", interm["mask_final"]),
    ]

    n = len(stages)
    plt.figure(figsize=(3.2 * n, 3.2))
    for i, (name, arr) in enumerate(stages, 1):
        ax = plt.subplot(1, n, i)
        if arr.dtype != np.uint8:
            im = ax.imshow(arr, interpolation="nearest")
            plt.colorbar(im, ax=ax, fraction=0.046, pad=0.04)
        else:
            ax.imshow(arr, cmap="gray", interpolation="nearest")
        ax.set_title(f"{title_prefix}\n{name}")
        ax.axis("off")
    plt.tight_layout()
    plt.show()


def _show_unit_grid(unit_masks: list[np.ndarray], *, title: str = "Generated unit cells"):
    N = len(unit_masks)
    ncols = int(math.ceil(math.sqrt(N)))
    nrows = int(math.ceil(N / ncols))

    plt.figure(figsize=(3 * ncols, 3 * nrows))
    for i, mask in enumerate(unit_masks):
        ax = plt.subplot(nrows, ncols, i + 1)
        ax.imshow(mask, cmap="gray", interpolation="nearest")
        ax.set_title(f"Unit {i}")
        ax.axis("off")
    plt.suptitle(title)
    plt.tight_layout()
    plt.show()


def _show_supercells(supercells: list[np.ndarray]):
    if not supercells:
        return
    plt.figure(figsize=(12, 3 * len(supercells)))
    for i, sc in enumerate(supercells):
        ax = plt.subplot(len(supercells), 1, i + 1)
        ax.imshow(sc, cmap="gray", interpolation="nearest")
        ax.set_title(f"Supercell {i} (1x4)")
        ax.axis("off")
    plt.tight_layout()
    plt.show()


# ============================================================
#  
# ============================================================

if __name__ == "__main__":
    cfg = MaskGenConfig()

    rng = np.random.default_rng(cfg.rng_seed)
    g = int(cfg.ctrl_grid)

    unit_masks: list[np.ndarray] = []
    intermediates_first = None

    lo, hi = cfg.ctrl_value_range
    for i in range(int(cfg.n_units_to_generate)):
        ctrl = rng.uniform(lo, hi, size=(g, g))

        if i == 0 and cfg.show_intermediate_for_first:
            intermediates_first = generate_unit_cell_mask(ctrl, cfg, return_intermediates=True)
            unit_masks.append(intermediates_first["mask_final"])
        else:
            unit_masks.append(generate_unit_cell_mask(ctrl, cfg))

    print(f"[INFO] Generated {len(unit_masks)} unit masks of shape {unit_masks[0].shape}")

    if intermediates_first is not None:
        _show_intermediates(intermediates_first, title_prefix="Unit 0")

    if cfg.show_all_units_grid:
        _show_unit_grid(unit_masks)

    supercells = []
    n = int(cfg.supercell_n_units)
    for start in range(0, len(unit_masks) - (n - 1), n):
        sc = build_supercell_1x4(unit_masks[start:start + n], cfg)
        supercells.append(sc)

    print(f"[INFO] Built {len(supercells)} supercells of shape {supercells[0].shape if supercells else None}")

    if cfg.show_supercells:
        _show_supercells(supercells)
```

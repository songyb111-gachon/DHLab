```python
"""
Z 해상도 수렴 테스트 (25, 20, 15, 10, 5 nm)
============================================================================

동일 마스크로 5가지 Z 해상도 시뮬레이션 후 5nm(기준) 대비 상대 오차 비교.
측벽 기울기 1° 메타서페이스.

2D 모니터 방식: 핵심 z-평면 4개에서만 필드를 기록하여 메모리 절약.

필요 패키지:
    pip install tidy3d numpy scipy matplotlib
"""

import os
import math
from typing import Dict, List

import numpy as np
from scipy import ndimage as ndi
from scipy.interpolate import RectBivariateSpline
import matplotlib.pyplot as plt

import tidy3d as td
from tidy3d import web
from tidy3d.plugins.polyslab import ComplexPolySlab

# ==================== 설정 ====================
RANDOM_SEED = 42
UNIT_PX = 128
SUPERCELL_NX = 4
SUPERCELL_NY = 4
NX = UNIT_PX * SUPERCELL_NX          # 512
NY = UNIT_PX * SUPERCELL_NY          # 512
PIXEL_SIZE_NM = 5.0
Z_TOTAL_NM = 3200

SIDEWALL_ANGLE_DEG = 1.0
SIDEWALL_ANGLE_RAD = math.radians(SIDEWALL_ANGLE_DEG)

WAVELENGTH_NM = 1050
EPS_SIO2 = 1.4585
EPS_NANOPILLAR = 5.8
SUBSTRATE_THICKNESS_NM = 350
METASURFACE_HEIGHT_NM = 350

Z_SUBSTRATE_BOTTOM_NM = 1050
Z_SUBSTRATE_TOP_NM = Z_SUBSTRATE_BOTTOM_NM + SUBSTRATE_THICKNESS_NM
Z_METASURFACE_TOP_NM = Z_SUBSTRATE_TOP_NM + METASURFACE_HEIGHT_NM
Z_SOURCE_NM = 3050
PML_THICKNESS_NM = 150
SIMULATION_TIME_FS = 500

Z_MONITOR_PLANES_NM = [
    Z_SUBSTRATE_TOP_NM,                                    # 기판-메타서페이스 경계
    (Z_SUBSTRATE_TOP_NM + Z_METASURFACE_TOP_NM) / 2,      # 메타서페이스 중간
    Z_METASURFACE_TOP_NM,                                  # 메타서페이스 상단
    Z_METASURFACE_TOP_NM + 200,                            # 자유공간 (위 200nm)
]
Z_PLANE_LABELS = ["sub-meta", "meta-mid", "meta-top", "free-space"]

CTRL_GRID = 3
SPLINE_ORDER = 2
THRESHOLD = 0.0
BUFFER_PX = 8
SURFACE_GAUSSIAN_SIGMA = 0.5
MFS_PX = 5
MIN_ISLAND_AREA_PX = 13
MASK_EDGE_SMOOTH_SIGMA = 8.0


class MaskGenerator:
    """메타서페이스 마스크 생성기 (유닛셀 + 슈퍼셀 freeform)"""

    def __init__(self, seed=RANDOM_SEED):
        self.rng = np.random.default_rng(seed)

    def _make_quadrant_surface(self, ctrl, qn):
        g = ctrl.shape[0]
        gx, gy = np.linspace(0, 1, g), np.linspace(0, 1, g)
        k = max(1, min(SPLINE_ORDER, g - 1))
        spline = RectBivariateSpline(gx, gy, ctrl, kx=k, ky=k)
        S = spline(np.linspace(0, 1, qn), np.linspace(0, 1, qn))
        if SURFACE_GAUSSIAN_SIGMA > 0:
            S = ndi.gaussian_filter(S, sigma=SURFACE_GAUSSIAN_SIGMA, mode="reflect")
        return S

    def _fourfold_symmetry(self, quadrant):
        left = np.fliplr(quadrant)
        top = np.concatenate([left, quadrant], axis=1)
        bottom = np.flipud(top)
        return np.concatenate([bottom, top], axis=0)

    def _apply_buffer(self, mask):
        m = mask.astype(np.uint8).copy()
        b = BUFFER_PX
        m[:b, :] = m[-b:, :] = m[:, :b] = m[:, -b:] = 0
        return m

    def _mfs_filter(self, mask, mfs_px=MFS_PX):
        m = mask.astype(bool)
        if mfs_px > 1:
            r = max(1, int(np.round(mfs_px / 2)))
            se = ndi.iterate_structure(ndi.generate_binary_structure(2, 1), r)
            m = ndi.binary_closing(ndi.binary_opening(m, se), se)
        m = m.astype(np.uint8)
        labeled, n = ndi.label(m)
        if n > 0:
            counts = np.bincount(labeled.ravel())
            for lab in np.where(counts < MIN_ISLAND_AREA_PX)[0]:
                if lab != 0:
                    m[labeled == lab] = 0
        return m

    def _smooth_edges(self, mask):
        if MASK_EDGE_SMOOTH_SIGMA <= 0:
            return mask
        m = mask.astype(bool)
        sdf = ndi.distance_transform_edt(m) - ndi.distance_transform_edt(~m)
        sdf_s = ndi.gaussian_filter(sdf, sigma=MASK_EDGE_SMOOTH_SIGMA, mode="reflect")
        return (sdf_s > 0).astype(np.uint8)

    def generate_unit_cell(self, eps_value):
        ctrl = self.rng.uniform(-1, 1, size=(CTRL_GRID, CTRL_GRID))
        quad = self._make_quadrant_surface(ctrl, UNIT_PX // 2)
        field = self._fourfold_symmetry(quad)
        mask = (field > THRESHOLD).astype(np.uint8)
        mask = self._apply_buffer(mask)
        mask = self._mfs_filter(mask)
        mask = self._smooth_edges(mask)
        mask = self._mfs_filter(mask, mfs_px=1)
        return np.where(mask > 0, eps_value, 1.0).astype(np.float32)

    def generate_supercell(self):
        supercell = np.ones((NY, NX), dtype=np.float32)
        for iy in range(SUPERCELL_NY):
            for ix in range(SUPERCELL_NX):
                self.rng.uniform(0, 1)
                unit = self.generate_unit_cell(EPS_NANOPILLAR)
                y0, y1 = iy * UNIT_PX, (iy + 1) * UNIT_PX
                x0, x1 = ix * UNIT_PX, (ix + 1) * UNIT_PX
                supercell[y0:y1, x0:x1] = unit
        return supercell

OUTPUT_DIR = "./z_resolution_convergence"
COMPARE_DZS = [25, 20, 15, 10]
REF_DZ = 5


def nm_to_um(x): return x * 1e-3


def build_simulation_with_dz(eps_2d: np.ndarray, dz_nm: float) -> td.Simulation:
    """지정된 Z 해상도로 시뮬레이션 구성. 2D 모니터 4개 사용."""
    dl_um = nm_to_um(PIXEL_SIZE_NM)
    dz_um = nm_to_um(dz_nm)
    pml_cells = max(1, int(PML_THICKNESS_NM / dz_nm))

    Lx = NX * dl_um
    Ly = NY * dl_um
    Lz = nm_to_um(Z_TOTAL_NM)

    z_sub_bottom = nm_to_um(Z_SUBSTRATE_BOTTOM_NM)
    z_sub_top = nm_to_um(Z_SUBSTRATE_TOP_NM)
    z_meta_top = nm_to_um(Z_METASURFACE_TOP_NM)
    z_source = nm_to_um(Z_SOURCE_NM)

    substrate = td.Structure(
        geometry=td.Box(
            center=(0, 0, (z_sub_bottom + z_sub_top) / 2),
            size=(td.inf, td.inf, z_sub_top - z_sub_bottom),
        ),
        medium=td.Medium(permittivity=EPS_SIO2),
        name="substrate",
    )
    structures: List[td.Structure] = [substrate]

    unique_eps = np.unique(eps_2d)
    for eps_val in unique_eps:
        if abs(eps_val - 1.0) < 0.01:
            continue
        mask = np.abs(eps_2d - eps_val) < 0.01
        if not mask.any():
            continue
        labeled, n_regions = ndi.label(mask)
        for region_id in range(1, n_regions + 1):
            region_mask = labeled == region_id
            ys, xs = np.where(region_mask)
            if len(xs) == 0:
                continue
            x_min, x_max = xs.min(), xs.max()
            y_min, y_max = ys.min(), ys.max()
            x0_um = (x_min - NX / 2) * dl_um
            x1_um = (x_max - NX / 2 + 1) * dl_um
            y0_um = (y_min - NY / 2) * dl_um
            y1_um = (y_max - NY / 2 + 1) * dl_um
            vertices = np.array(
                [[x0_um, y0_um], [x1_um, y0_um], [x1_um, y1_um], [x0_um, y1_um]]
            )
            slab_bounds = (z_sub_top, z_meta_top)
            cps = ComplexPolySlab(
                vertices=vertices,
                slab_bounds=slab_bounds,
                axis=2,
                sidewall_angle=SIDEWALL_ANGLE_RAD,
                reference_plane="bottom",
            )
            mat = td.Medium(permittivity=float(eps_val))
            for sub in cps.sub_polyslabs:
                structures.append(
                    td.Structure(
                        geometry=sub, medium=mat, name=f"pillar_{len(structures)}"
                    )
                )

    freq0 = td.C_0 / nm_to_um(WAVELENGTH_NM)
    fwidth = freq0 * 0.1
    source = td.PlaneWave(
        center=(0, 0, z_source),
        size=(td.inf, td.inf, 0),
        source_time=td.GaussianPulse(freq0=freq0, fwidth=fwidth),
        direction="-",
        pol_angle=0.0,
        name="source",
    )

    monitors = []
    for z_nm, label in zip(Z_MONITOR_PLANES_NM, Z_PLANE_LABELS):
        monitors.append(
            td.FieldMonitor(
                center=(0, 0, nm_to_um(z_nm)),
                size=(Lx * 0.99, Ly * 0.99, 0),
                freqs=[freq0],
                name=f"field_{label}",
            )
        )

    boundary_spec = td.BoundarySpec(
        x=td.Boundary(plus=td.Periodic(), minus=td.Periodic()),
        y=td.Boundary(plus=td.Periodic(), minus=td.Periodic()),
        z=td.Boundary(
            plus=td.PML(num_layers=pml_cells), minus=td.PML(num_layers=pml_cells)
        ),
    )

    grid_spec = td.GridSpec(
        grid_x=td.UniformGrid(dl=dl_um),
        grid_y=td.UniformGrid(dl=dl_um),
        grid_z=td.UniformGrid(dl=dz_um),
    )

    return td.Simulation(
        center=(0, 0, Lz / 2),
        size=(Lx, Ly, Lz),
        grid_spec=grid_spec,
        structures=structures,
        sources=[source],
        monitors=monitors,
        boundary_spec=boundary_spec,
        run_time=SIMULATION_TIME_FS * 1e-15,
    )


def extract_H_2d(sim_data, monitor_name: str, freq_idx: int = 0) -> np.ndarray:
    """2D 모니터에서 H 필드 추출. 반환 shape: (Nx, Ny, 3) complex"""
    fd = sim_data[monitor_name]
    components = []
    for comp in ["Hx", "Hy", "Hz"]:
        d = getattr(fd, comp)
        if "f" in d.dims:
            d = d.isel(f=freq_idx)
        components.append(np.squeeze(d.values))
    return np.stack(components, axis=-1)


def compute_errors(H_ref: np.ndarray, H_test: np.ndarray) -> Dict[str, float]:
    """두 2D H 필드 간 오차 계산"""
    diff = np.abs(H_ref - H_test)
    denom_safe = np.where(np.abs(H_ref) > 1e-12, np.abs(H_ref), 1e-12)
    rel_rms = float(np.sqrt(np.mean((diff / denom_safe) ** 2)))
    l2_norm = float(
        np.linalg.norm(H_ref - H_test) / (np.linalg.norm(H_ref) + 1e-12)
    )
    return {"relative_RMS": rel_rms, "L2_normalized": l2_norm}


def main():
    os.makedirs(OUTPUT_DIR, exist_ok=True)
    n_planes = len(Z_PLANE_LABELS)
    mon_names = [f"field_{lb}" for lb in Z_PLANE_LABELS]

    print("=" * 70)
    print("Z 해상도 수렴 테스트 — 2D 모니터 방식 (메모리 절약)")
    print(f"  모니터 평면: {Z_PLANE_LABELS}")
    print(f"  비교 해상도: {COMPARE_DZS} nm  vs  기준 {REF_DZ} nm")
    print(f"  시뮬레이션 시간: {SIMULATION_TIME_FS} fs")
    print("=" * 70)

    mask_gen = MaskGenerator(seed=RANDOM_SEED)
    eps_2d = mask_gen.generate_supercell()

    # ── 1단계: 기준(5nm) 시뮬레이션 ──
    print(f"\n>>> [1/{len(COMPARE_DZS)+1}] 기준 DZ={REF_DZ}nm <<<")
    sim_ref = build_simulation_with_dz(eps_2d, REF_DZ)
    nz_ref = int(Z_TOTAL_NM / REF_DZ)
    print(f"  NZ={nz_ref}, voxels={NX*NY*nz_ref/1e6:.1f}M")
    sim_data_ref = web.run(sim_ref, task_name=f"z_conv_dz{REF_DZ}nm")

    H_ref_planes = {}
    for mn in mon_names:
        H_ref_planes[mn] = extract_H_2d(sim_data_ref, mn)
    del sim_data_ref

    fig, axes = plt.subplots(1, n_planes, figsize=(4 * n_planes, 4))
    for j, (mn, lb) in enumerate(zip(mon_names, Z_PLANE_LABELS)):
        H_mag = np.sum(np.abs(H_ref_planes[mn]) ** 2, axis=-1)
        im = axes[j].imshow(H_mag.T, origin="lower", cmap="hot")
        axes[j].set_title(f"|H|² ref @ {lb}")
        plt.colorbar(im, ax=axes[j], fraction=0.046)
    plt.suptitle(f"Reference fields (DZ={REF_DZ}nm)", fontsize=13)
    plt.tight_layout()
    plt.savefig(os.path.join(OUTPUT_DIR, "reference.png"), dpi=150)
    plt.show()
    print("  기준 필드 저장 완료")

    # ── 2단계: 각 해상도별 시뮬레이션 → 즉시 비교 ──
    all_errors: Dict[int, Dict[str, Dict[str, float]]] = {}

    for i, dz_nm in enumerate(COMPARE_DZS):
        step = i + 2
        print(f"\n>>> [{step}/{len(COMPARE_DZS)+1}] DZ={dz_nm}nm <<<")
        sim = build_simulation_with_dz(eps_2d, dz_nm)
        nz = int(Z_TOTAL_NM / dz_nm)
        print(f"  NZ={nz}, voxels={NX*NY*nz/1e6:.1f}M")
        sim_data = web.run(sim, task_name=f"z_conv_dz{dz_nm}nm")

        plane_errors = {}
        H_test_planes = {}
        for mn, lb in zip(mon_names, Z_PLANE_LABELS):
            H_test = extract_H_2d(sim_data, mn)
            H_test_planes[mn] = H_test
            plane_errors[lb] = compute_errors(H_ref_planes[mn], H_test)
            print(
                f"    {lb}: rel_RMS={plane_errors[lb]['relative_RMS']:.4f}, "
                f"L2={plane_errors[lb]['L2_normalized']:.4f}"
            )
        all_errors[dz_nm] = plane_errors
        del sim_data

        # ── 즉시 시각화 ──
        done_dzs = COMPARE_DZS[: i + 1]

        fig, axes = plt.subplots(2, n_planes, figsize=(4 * n_planes, 8))
        fig.suptitle(
            f"Convergence — {len(done_dzs)}/{len(COMPARE_DZS)} done "
            f"(latest: {dz_nm}nm)",
            fontsize=13,
        )

        for j, (mn, lb) in enumerate(zip(mon_names, Z_PLANE_LABELS)):
            ax_bar = axes[0, j]
            vals = [all_errors[d][lb]["L2_normalized"] for d in done_dzs]
            colors = ["steelblue"] * len(vals)
            colors[-1] = "orange"
            ax_bar.bar([str(d) for d in done_dzs], vals, color=colors, edgecolor="k")
            ax_bar.set_title(f"L2 @ {lb}")
            ax_bar.set_xlabel("DZ (nm)")
            if len(vals) > 1 and max(vals) / (min(vals) + 1e-30) > 5:
                ax_bar.set_yscale("log")

            ax_diff = axes[1, j]
            H_mag_ref = np.sum(np.abs(H_ref_planes[mn]) ** 2, axis=-1)
            H_mag_test = np.sum(np.abs(H_test_planes[mn]) ** 2, axis=-1)
            diff_map = np.abs(H_mag_ref - H_mag_test)
            im = ax_diff.imshow(diff_map.T, origin="lower", cmap="magma")
            ax_diff.set_title(f"Δ|H|² {dz_nm}nm @ {lb}")
            plt.colorbar(im, ax=ax_diff, fraction=0.046)

        plt.tight_layout()
        plt.savefig(os.path.join(OUTPUT_DIR, f"step{step}_dz{dz_nm}nm.png"), dpi=150)
        plt.show()

    # ── 최종 요약 테이블 출력 ──
    print("\n" + "=" * 70)
    header = f"{'DZ(nm)':>8}"
    for lb in Z_PLANE_LABELS:
        header += f"  {lb+' RMS':>14}  {lb+' L2':>12}"
    print(header)
    print("-" * len(header))
    for dz_nm in COMPARE_DZS:
        row = f"{dz_nm:>8}"
        for lb in Z_PLANE_LABELS:
            e = all_errors[dz_nm][lb]
            row += f"  {e['relative_RMS']:>14.5f}  {e['L2_normalized']:>12.5f}"
        print(row)
    print("=" * 70)

    np.savez(
        os.path.join(OUTPUT_DIR, "convergence_results.npz"),
        eps_2d=eps_2d,
        plane_labels=Z_PLANE_LABELS,
        dz_compared=np.array(COMPARE_DZS),
        **{
            f"H_ref_{lb}": H_ref_planes[mn]
            for mn, lb in zip(mon_names, Z_PLANE_LABELS)
        },
        **{
            f"err_rms_{lb}": np.array(
                [all_errors[d][lb]["relative_RMS"] for d in COMPARE_DZS]
            )
            for lb in Z_PLANE_LABELS
        },
        **{
            f"err_l2_{lb}": np.array(
                [all_errors[d][lb]["L2_normalized"] for d in COMPARE_DZS]
            )
            for lb in Z_PLANE_LABELS
        },
    )
    print(f"\n결과 저장: {OUTPUT_DIR}/convergence_results.npz")
    print("완료!")


if __name__ == "__main__":
    main()
```
